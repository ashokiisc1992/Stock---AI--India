```python
# ============================================
# TECHNICAL INDICATORS CALCULATOR
# ============================================

def calculate_indicators(df):
    """
    Calculate all technical indicators for a stock
    Input: OHLCV dataframe from yfinance
    Output: Same dataframe with indicators added
    """
    data = df.copy()

    # ── EMA 20, 50, 200 ──────────────────────
    data['EMA20']  = data['Close'].ewm(span=20,  adjust=False).mean()
    data['EMA50']  = data['Close'].ewm(span=50,  adjust=False).mean()
    data['EMA200'] = data['Close'].ewm(span=200, adjust=False).mean()

    # ── RSI (14 period) ──────────────────────
    delta     = data['Close'].diff()
    gain      = delta.where(delta > 0, 0)
    loss      = -delta.where(delta < 0, 0)
    avg_gain  = gain.ewm(span=14, adjust=False).mean()
    avg_loss  = loss.ewm(span=14, adjust=False).mean()
    rs        = avg_gain / avg_loss
    data['RSI'] = 100 - (100 / (1 + rs))

    # ── MACD (12, 26, 9) ─────────────────────
    ema12          = data['Close'].ewm(span=12, adjust=False).mean()
    ema26          = data['Close'].ewm(span=26, adjust=False).mean()
    data['MACD']   = ema12 - ema26
    data['Signal'] = data['MACD'].ewm(span=9, adjust=False).mean()
    data['MACD_Hist'] = data['MACD'] - data['Signal']

    # ── ADX (14 period) ──────────────────────
    high        = data['High']
    low         = data['Low']
    close       = data['Close']

    # True Range
    tr1         = high - low
    tr2         = (high - close.shift()).abs()
    tr3         = (low  - close.shift()).abs()
    tr          = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)

    # Directional Movement
    dm_plus     = high.diff()
    dm_minus    = -low.diff()
    dm_plus     = dm_plus.where((dm_plus > dm_minus) & (dm_plus > 0), 0)
    dm_minus    = dm_minus.where((dm_minus > dm_plus) & (dm_minus > 0), 0)

    # Smooth with EMA
    atr         = tr.ewm(span=14,       adjust=False).mean()
    di_plus     = 100 * dm_plus.ewm(span=14,  adjust=False).mean() / atr
    di_minus    = 100 * dm_minus.ewm(span=14, adjust=False).mean() / atr
    dx          = 100 * (di_plus - di_minus).abs() / (di_plus + di_minus)
    data['ADX'] = dx.ewm(span=14, adjust=False).mean()
    data['DI_Plus']  = di_plus
    data['DI_Minus'] = di_minus

    # ── Volume Moving Average ─────────────────
    data['Vol_MA20'] = data['Volume'].rolling(20).mean()
    data['Vol_Ratio'] = data['Volume'] / data['Vol_MA20']

    return data


# Test on one stock
print("Testing indicators on PERSISTENT...")
test_df = calculate_indicators(price_data['PERSISTENT'])

# Show latest values
latest = test_df.iloc[-1]
print(f"\nPERSISTENT — Latest Indicator Values:")
print(f"  Close    : ₹{latest['Close']:.2f}")
print(f"  EMA20    : ₹{latest['EMA20']:.2f}")
print(f"  EMA50    : ₹{latest['EMA50']:.2f}")
print(f"  EMA200   : ₹{latest['EMA200']:.2f}")
print(f"  RSI      : {latest['RSI']:.2f}")
print(f"  MACD     : {latest['MACD']:.2f}")
print(f"  Signal   : {latest['Signal']:.2f}")
print(f"  MACD Hist: {latest['MACD_Hist']:.2f}")
print(f"  ADX      : {latest['ADX']:.2f}")
print(f"  Vol Ratio: {latest['Vol_Ratio']:.2f}x avg")
```

```python
# ============================================
# VOLUME PROFILE CALCULATOR
# ============================================

def calculate_volume_profile(df, lookback_days=252, bins=30):
    """
    Calculate volume profile for a stock
    lookback_days: how many days to include (252 = 1 year)
    bins: number of price levels
    
    Returns:
    - poc_price    : Point of Control (highest volume price)
    - val          : Value Area Low (bottom of 70% volume zone)
    - vah          : Value Area High (top of 70% volume zone)
    - is_bell      : True if bell shape detected
    - breakout     : BREAKOUT_UP / BREAKOUT_DOWN / INSIDE / LVN_UP
    """
    # Use recent data for volume profile
    recent    = df.tail(lookback_days).copy()
    
    price_min = recent['Low'].min()
    price_max = recent['High'].max()
    
    if price_max <= price_min:
        return None
    
    # Create price buckets
    price_bins       = np.linspace(price_min, price_max, bins + 1)
    volume_at_price  = np.zeros(bins)
    
    # Distribute volume across price range for each candle
    for _, row in recent.iterrows():
        candle_low  = row['Low']
        candle_high = row['High']
        candle_vol  = row['Volume']
        candle_range = candle_high - candle_low
        
        if candle_range == 0:
            continue
        
        for i in range(bins):
            bin_low  = price_bins[i]
            bin_high = price_bins[i + 1]
            
            overlap_low  = max(candle_low,  bin_low)
            overlap_high = min(candle_high, bin_high)
            
            if overlap_high > overlap_low:
                overlap_pct          = (overlap_high - overlap_low) / candle_range
                volume_at_price[i]  += candle_vol * overlap_pct
    
    # ── FIND POC ──────────────────────────────
    poc_idx   = np.argmax(volume_at_price)
    poc_price = (price_bins[poc_idx] + price_bins[poc_idx + 1]) / 2
    
    # ── FIND VALUE AREA (70% of volume) ──────
    total_vol  = volume_at_price.sum()
    target_vol = total_vol * 0.70
    
    lower_idx  = poc_idx
    upper_idx  = poc_idx
    va_vol     = volume_at_price[poc_idx]
    
    while va_vol < target_vol:
        can_expand_up   = upper_idx + 1 < bins
        can_expand_down = lower_idx - 1 >= 0
        
        expand_up   = volume_at_price[upper_idx + 1] if can_expand_up   else 0
        expand_down = volume_at_price[lower_idx - 1] if can_expand_down else 0
        
        if not can_expand_up and not can_expand_down:
            break
        
        if expand_up >= expand_down and can_expand_up:
            upper_idx += 1
            va_vol    += expand_up
        elif can_expand_down:
            lower_idx -= 1
            va_vol    += expand_down
        else:
            break
    
    val = (price_bins[lower_idx] + price_bins[lower_idx + 1]) / 2
    vah = (price_bins[upper_idx] + price_bins[upper_idx + 1]) / 2
    
    # ── DETECT BELL SHAPE ────────────────────
    max_vol    = volume_at_price.max()
    edge_vol   = (volume_at_price[:3].mean() + volume_at_price[-3:].mean()) / 2
    is_bell    = max_vol > edge_vol * 2.5
    
    # ── DETECT LOW VOLUME NODES ──────────────
    avg_vol    = volume_at_price.mean()
    lvn_above  = []
    for i in range(poc_idx + 1, bins):
        if volume_at_price[i] < avg_vol * 0.4:
            lvn_above.append(
                (price_bins[i] + price_bins[i + 1]) / 2
            )
    
    # ── BREAKOUT DETECTION ───────────────────
    current_price = df['Close'].iloc[-1]
    
    if current_price > vah * 1.02:
        breakout_status = 'BREAKOUT_UP'
    elif current_price < val * 0.98:
        breakout_status = 'BREAKOUT_DOWN'
    elif current_price > vah * 0.98:
        breakout_status = 'AT_RESISTANCE'
    elif lvn_above and current_price > poc_price:
        breakout_status = 'LVN_ABOVE'
    else:
        breakout_status = 'INSIDE'
    
    return {
        'poc_price'      : round(poc_price, 2),
        'val'            : round(val,       2),
        'vah'            : round(vah,       2),
        'is_bell'        : is_bell,
        'breakout_status': breakout_status,
        'current_price'  : round(current_price, 2),
        'lvn_above'      : lvn_above[:3],
        'price_vs_poc'   : round(
            (current_price - poc_price) / poc_price * 100, 2
        ),
        'price_vs_vah'   : round(
            (current_price - vah) / vah * 100, 2
        ),
    }


# Test on PERSISTENT and DHANUKA
for symbol in ['PERSISTENT', 'DHANUKA', 'BEL', 'MUTHOOTFIN']:
    vp = calculate_volume_profile(indicator_data[symbol])
    print(f"\n{symbol}:")
    print(f"  Current Price  : ₹{vp['current_price']}")
    print(f"  POC            : ₹{vp['poc_price']}")
    print(f"  Value Area Low : ₹{vp['val']}")
    print(f"  Value Area High: ₹{vp['vah']}")
    print(f"  Bell Shape     : {vp['is_bell']}")
    print(f"  Status         : {vp['breakout_status']}")
    print(f"  Price vs POC   : {vp['price_vs_poc']}%")
    print(f"  Price vs VAH   : {vp['price_vs_vah']}%")
```

```python
# ============================================
# SETUP 1 — MOMENTUM SCORE
# ============================================

def score_momentum(df, vp):
    """
    Score momentum continuation setup
    Stock in uptrend with strong momentum
    Returns 0-100
    """
    scores  = {}
    latest  = df.iloc[-1]
    close   = latest['Close']

    # ── EMA ALIGNMENT (30 pts) ──
    ema20   = latest['EMA20']
    ema50   = latest['EMA50']
    ema200  = latest['EMA200']

    if close > ema20 > ema50 > ema200:
        scores['EMA'] = 30    # Perfect alignment ✅
    elif close > ema50 > ema200:
        scores['EMA'] = 20    # Above 50 and 200
    elif close > ema200:
        scores['EMA'] = 10    # Only above 200
    else:
        scores['EMA'] = 0     # Below all EMAs ❌

    # ── RSI ZONE (20 pts) ──
    rsi     = latest['RSI']

    if 50 <= rsi <= 70:
        scores['RSI'] = 20    # Ideal momentum zone ✅
    elif 45 <= rsi < 50:
        scores['RSI'] = 12    # Just below ideal
    elif 70 < rsi <= 80:
        scores['RSI'] = 8     # Slightly overbought
    elif rsi > 80:
        scores['RSI'] = 0     # Overbought ❌
    else:
        scores['RSI'] = 0     # Below 45 = no momentum

    # ── MACD BULLISH (20 pts) ──
    macd      = latest['MACD']
    signal    = latest['Signal']
    hist      = latest['MACD_Hist']

    if macd > signal and hist > 0 and macd > 0:
        scores['MACD'] = 20   # Fully bullish ✅
    elif macd > signal and hist > 0:
        scores['MACD'] = 15   # Bullish crossover
    elif macd > signal:
        scores['MACD'] = 10   # Above signal
    else:
        scores['MACD'] = 0    # Bearish ❌

    # ── ADX STRENGTH (15 pts) ──
    adx       = latest['ADX']
    di_plus   = latest['DI_Plus']
    di_minus  = latest['DI_Minus']

    if adx > 25 and di_plus > di_minus:
        scores['ADX'] = 15    # Strong uptrend ✅
    elif adx > 20 and di_plus > di_minus:
        scores['ADX'] = 10    # Moderate uptrend
    elif adx > 25:
        scores['ADX'] = 5     # Strong but downtrend
    else:
        scores['ADX'] = 0     # Weak trend ❌

    # ── VOLUME PROFILE (15 pts) ──
    if vp:
        status = vp['breakout_status']
        if status == 'BREAKOUT_UP':
            scores['VP'] = 15  # Above value area ✅
        elif status == 'AT_RESISTANCE':
            scores['VP'] = 10  # Testing breakout
        elif status == 'LVN_ABOVE':
            scores['VP'] = 12  # In thin area
        elif status == 'INSIDE':
            scores['VP'] = 5   # Inside value area
        else:
            scores['VP'] = 0   # Below value area ❌
    else:
        scores['VP'] = 5

    total = sum(scores.values())
    return total, scores


# ============================================
# SETUP 2 — REVERSAL SCORE (DIVERGENCE)
# ============================================

def score_reversal(df, vp, lookback=20):
    """
    Score bullish divergence reversal setup
    Price making lower lows but indicators higher lows
    Returns 0-100
    """
    scores  = {}

    # Need enough data
    if len(df) < lookback + 10:
        return 0, {}

    recent  = df.tail(lookback + 10)

    # ── RSI DIVERGENCE (35 pts) ──
    # Price making lower low but RSI making higher low
    price_recent  = recent['Close'].tail(lookback)
    rsi_recent    = recent['RSI'].tail(lookback)

    price_min_idx = price_recent.idxmin()
    price_min2    = price_recent[
        price_recent.index < price_min_idx
    ].min() if len(price_recent[
        price_recent.index < price_min_idx
    ]) > 0 else None

    rsi_at_low    = rsi_recent[price_min_idx]
    rsi_at_prev   = rsi_recent[
        price_recent.idxmin()
    ] if price_min2 else None

    latest_rsi    = df['RSI'].iloc[-1]
    latest_close  = df['Close'].iloc[-1]

    # Check if RSI is oversold and rising
    if latest_rsi < 35:
        scores['RSI_Level'] = 20   # Oversold ✅
    elif latest_rsi < 45:
        scores['RSI_Level'] = 10
    else:
        scores['RSI_Level'] = 0

    # Check RSI divergence
    # RSI making higher low while price lower
    rsi_5d_min    = df['RSI'].tail(5).min()
    rsi_20d_min   = df['RSI'].tail(20).min()
    price_5d_min  = df['Close'].tail(5).min()
    price_20d_min = df['Close'].tail(20).min()

    if (price_5d_min <= price_20d_min and
            rsi_5d_min > rsi_20d_min):
        scores['RSI_Divergence'] = 15  # Classic divergence ✅
    else:
        scores['RSI_Divergence'] = 0

    # ── MACD DIVERGENCE (35 pts) ──
    latest_hist   = df['MACD_Hist'].iloc[-1]
    prev_hist     = df['MACD_Hist'].iloc[-5]

    # MACD histogram turning from negative to positive
    if latest_hist > 0 and prev_hist < 0:
        scores['MACD_Cross'] = 20  # Fresh crossover ✅
    elif latest_hist > prev_hist and latest_hist < 0:
        scores['MACD_Rising'] = 15  # Rising but still negative
    else:
        scores['MACD_Cross']  = 0
        scores['MACD_Rising'] = 0

    # MACD divergence
    macd_5d_min   = df['MACD_Hist'].tail(5).min()
    macd_20d_min  = df['MACD_Hist'].tail(20).min()

    if (price_5d_min <= price_20d_min and
            macd_5d_min > macd_20d_min):
        scores['MACD_Divergence'] = 15  # MACD divergence ✅
    else:
        scores['MACD_Divergence'] = 0

    # ── ADX DECLINING (15 pts) ──
    adx_latest = df['ADX'].iloc[-1]
    adx_10d    = df['ADX'].iloc[-10]

    if adx_latest < adx_10d and adx_latest > 20:
        scores['ADX'] = 15   # Downtrend weakening ✅
    elif adx_latest < adx_10d:
        scores['ADX'] = 8
    else:
        scores['ADX'] = 0

    # ── VOLUME DRYING UP (15 pts) ──
    vol_ratio = df['Vol_Ratio'].iloc[-1]

    if vol_ratio < 0.6:
        scores['Volume'] = 15  # Very low volume = sellers exhausted ✅
    elif vol_ratio < 0.8:
        scores['Volume'] = 8
    else:
        scores['Volume'] = 0

    total = sum(scores.values())
    return min(total, 100), scores


# ============================================
# SETUP 3 — BREAKOUT SCORE
# ============================================

def score_breakout(df, vp, lookback=60):
    """
    Score range breakout setup
    Stock consolidating then breaking out with volume
    Returns 0-100
    """
    scores = {}

    if len(df) < lookback:
        return 0, {}

    recent        = df.tail(lookback)
    latest        = df.iloc[-1]
    close         = latest['Close']

    # ── BELL SHAPE / CONSOLIDATION (25 pts) ──
    if vp and vp['is_bell']:
        scores['Bell'] = 25   # Bell shape confirmed ✅
    else:
        scores['Bell'] = 0

    # ── PRICE AT BREAKOUT LEVEL (20 pts) ──
    if vp:
        status = vp['breakout_status']
        if status == 'BREAKOUT_UP':
            scores['Position'] = 20   # Already broke out ✅
        elif status == 'AT_RESISTANCE':
            scores['Position'] = 15   # Testing breakout level
        elif status == 'LVN_ABOVE':
            scores['Position'] = 10
        else:
            scores['Position'] = 0
    else:
        scores['Position'] = 0

    # ── VOLUME ON BREAKOUT (25 pts) ──
    vol_ratio = latest['Vol_Ratio']

    if vol_ratio > 2.5:
        scores['Volume'] = 25   # Massive volume ✅
    elif vol_ratio > 2.0:
        scores['Volume'] = 20
    elif vol_ratio > 1.5:
        scores['Volume'] = 12
    elif vol_ratio > 1.0:
        scores['Volume'] = 6
    else:
        scores['Volume'] = 0    # No volume = weak breakout

    # ── EMA COMPRESSION (15 pts) ──
    ema20  = latest['EMA20']
    ema50  = latest['EMA50']

    # EMAs close together = consolidation
    ema_gap = abs(ema20 - ema50) / ema50 * 100

    if ema_gap < 2:
        scores['EMA_Compression'] = 15   # Very tight ✅
    elif ema_gap < 5:
        scores['EMA_Compression'] = 8
    else:
        scores['EMA_Compression'] = 0

    # ── ADX RISING FROM LOW (15 pts) ──
    adx_now  = latest['ADX']
    adx_10d  = df['ADX'].iloc[-10]
    adx_20d  = df['ADX'].iloc[-20]

    # ADX was low (consolidation) and now rising (breakout starting)
    if adx_20d < 20 and adx_now > adx_10d:
        scores['ADX'] = 15   # Classic breakout ADX ✅
    elif adx_now > adx_10d > adx_20d:
        scores['ADX'] = 8    # ADX rising
    else:
        scores['ADX'] = 0

    total = sum(scores.values())
    return min(total, 100), scores


print("All three setup scorers loaded ✅")
```

```python
def detect_dynamic_consolidation(df, min_days=10, max_days=500):
    """
    Dynamically find consolidation range regardless of duration
    
    Logic:
    1. Start with last 10 days to find initial range
    2. Expand backwards as long as price stays in range
    3. Stop when price breaks outside the range
    4. That defines the consolidation period
    """
    
    close   = df['Close']
    high    = df['High']
    low     = df['Low']
    n       = len(df)
    
    # Start with last min_days to define initial range
    initial = df.tail(min_days)
    range_high = initial['High'].max()
    range_low  = initial['Low'].min()
    
    # Expand backwards
    consol_start = n - min_days
    
    for i in range(n - min_days - 1, max(0, n - max_days), -1):
        bar_high = df['High'].iloc[i]
        bar_low  = df['Low'].iloc[i]
        
        # If this bar is within the range → extend consolidation
        if bar_low >= range_low * 0.97 and bar_high <= range_high * 1.03:
            consol_start = i
            # Optionally tighten the range
            range_high   = max(range_high, bar_high)
            range_low    = min(range_low,  bar_low)
        else:
            # Price went outside range → consolidation started here
            break
    
    consol_days  = n - consol_start
    current_price = close.iloc[-1]
    
    # Is price breaking out now?
    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98
    
    # Range width
    range_pct     = (range_high - range_low) / range_low * 100
    
    # Was it sideways? (slope check on consolidation period)
    consol_prices = close.iloc[consol_start:].values
    x             = np.arange(len(consol_prices))
    if len(consol_prices) > 2:
        slope         = np.polyfit(x, consol_prices, 1)[0]
        norm_slope    = abs(slope) / consol_prices.mean() * 100
        is_sideways   = norm_slope < 0.08
    else:
        is_sideways   = False
        norm_slope    = 999

    # Count touches of resistance and support
    consol_df     = df.iloc[consol_start:]
    res_touches   = (consol_df['High'] >= range_high * 0.98).sum()
    sup_touches   = (consol_df['Low']  <= range_low  * 1.02).sum()
    
    return {
        'found'               : consol_days >= min_days and is_sideways,
        'consol_days'         : consol_days,
        'range_high'          : round(range_high,  2),
        'range_low'           : round(range_low,   2),
        'range_pct'           : round(range_pct,   2),
        'is_sideways'         : is_sideways,
        'norm_slope'          : round(norm_slope,  4),
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above,   2),
        'resistance_touches'  : int(res_touches),
        'support_touches'     : int(sup_touches),
        'current_price'       : round(current_price, 2),
        'breakout_volume'     : round(
            df['Vol_Ratio'].tail(5).max(), 2
        ),
    }


# Test on all stocks
print("=== DYNAMIC CONSOLIDATION DETECTION ===\n")
for symbol in ['MUTHOOTFIN', 'BEL', 'SUNPHARMA',
               'IPCALAB', 'ABSLAMC', 'DHANUKA',
               'PERSISTENT', 'VINATIORGA']:
    df     = indicator_data[symbol]
    result = detect_dynamic_consolidation(df)
    
    print(f"{symbol:15} "
          f"Found: {str(result['found']):5}  "
          f"Days: {result['consol_days']:4}  "
          f"Range: {result['range_pct']:5.1f}%  "
          f"Sideways: {str(result['is_sideways']):5}  "
          f"Breaking: {str(result['is_breaking_out']):5}  "
          f"Vol: {result['breakout_volume']:.1f}x")
```

```python
def find_best_consolidation(df):
    """
    Check multiple timeframes for consolidation
    Returns the best (most recent + most valid) one found
    """
    timeframes = [
        {'min': 10,  'max': 30,  'label': 'Short  (2-4 weeks)'},
        {'min': 30,  'max': 90,  'label': 'Medium (1-3 months)'},
        {'min': 90,  'max': 250, 'label': 'Long   (3-12 months)'},
        {'min': 250, 'max': 500, 'label': 'Major  (1-2 years)'},
    ]

    results = []

    for tf in timeframes:
        result = detect_dynamic_consolidation(
            df,
            min_days = tf['min'],
            max_days = tf['max']
        )
        result['label'] = tf['label']

        # Valid consolidation criteria
        is_valid = (
            result['found']
            and result['range_pct'] < 30        # Tight range
            and result['resistance_touches'] >= 2 # Multiple touches
            and result['support_touches'] >= 2
        )
        result['is_valid'] = is_valid
        results.append(result)

    # Return best valid result (prefer longer consolidation)
    valid = [r for r in results if r['is_valid']]

    if valid:
        # Prefer longer consolidation — bigger base = bigger move
        best = max(valid, key=lambda x: x['consol_days'])
        return best

    # No valid consolidation found
    return {
        'found'          : False,
        'is_valid'       : False,
        'consol_days'    : 0,
        'range_pct'      : 0,
        'is_breaking_out': False,
        'label'          : 'None',
        'range_high'     : 0,
        'range_low'      : 0,
        'current_price'  : df['Close'].iloc[-1],
        'breakout_volume': df['Vol_Ratio'].tail(5).max(),
        'resistance_touches': 0,
        'support_touches'   : 0,
        'pct_above'         : 0,
    }


# Test on all stocks
print("=== MULTI-TIMEFRAME CONSOLIDATION DETECTION ===\n")
print(f"{'Symbol':15} {'Valid':6} {'Days':6} "
      f"{'Range%':8} {'Timeframe':25} "
      f"{'Breaking':9} {'Vol':6}")
print("─" * 80)

for symbol in ['MUTHOOTFIN', 'BEL', 'SUNPHARMA',
               'IPCALAB', 'ABSLAMC', 'DHANUKA',
               'PERSISTENT', 'VINATIORGA', 'TITAN',
               'HAL', 'GARFIBRES', 'TCS']:
    df     = indicator_data[symbol]
    result = find_best_consolidation(df)

    print(f"{symbol:15} "
          f"{str(result['is_valid']):6} "
          f"{result['consol_days']:6} "
          f"{result['range_pct']:7.1f}% "
          f"  {result['label']:25} "
          f"{str(result['is_breaking_out']):9} "
          f"{result['breakout_volume']:.1f}x")
```

```python
def detect_consolidation_dynamic(df, 
                                  min_days=10,
                                  max_range_pct=30,
                                  slope_threshold=0.06):
    """
    Fully dynamic consolidation detection
    No fixed timeframes — finds whatever consolidation exists
    
    Algorithm:
    1. Start from today
    2. Expand backwards one day at a time
    3. Keep expanding as long as:
       - New bar fits within the growing range
       - Range stays tight (< max_range_pct)
       - Slope stays flat
    4. Stop when any condition breaks
    5. That defines the consolidation period
    """
    
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)
    
    if n < min_days + 5:
        return {'found': False, 'reason': 'Insufficient data'}
    
    # Initialize with last 3 days
    range_high = high.iloc[-3:].max()
    range_low  = low.iloc[-3:].min()
    
    consol_start = n - 3
    last_valid   = n - 3
    
    # Expand backwards
    for i in range(n - 4, max(0, n - 1000), -1):
        
        bar_high   = high.iloc[i]
        bar_low    = low.iloc[i]
        
        # Tentatively expand range
        new_high   = max(range_high, bar_high)
        new_low    = min(range_low,  bar_low)
        new_range  = (new_high - new_low) / new_low * 100
        
        # Stop if range becomes too wide
        if new_range > max_range_pct:
            break
        
        # Check slope of consolidation so far
        consol_prices = close.iloc[i:].values
        if len(consol_prices) > 5:
            x          = np.arange(len(consol_prices))
            slope      = np.polyfit(x, consol_prices, 1)[0]
            norm_slope = abs(slope) / consol_prices.mean() * 100
            
            # Stop if trending too much
            if norm_slope > slope_threshold:
                break
        
        # Valid bar — extend consolidation
        range_high   = new_high
        range_low    = new_low
        consol_start = i
    
    consol_days   = n - consol_start
    current_price = close.iloc[-1]
    
    # Minimum duration check
    if consol_days < min_days:
        return {
            'found'       : False,
            'consol_days' : consol_days,
            'reason'      : f'Too short: {consol_days} days'
        }
    
    # Range width
    range_pct     = (range_high - range_low) / range_low * 100
    
    # Final slope check on full consolidation
    consol_prices = close.iloc[consol_start:].values
    x             = np.arange(len(consol_prices))
    slope         = np.polyfit(x, consol_prices, 1)[0]
    norm_slope    = abs(slope) / consol_prices.mean() * 100
    is_sideways   = norm_slope < slope_threshold
    
    # Count touches
    consol_df     = df.iloc[consol_start:]
    res_touches   = int((consol_df['High'] >= range_high * 0.98).sum())
    sup_touches   = int((consol_df['Low']  <= range_low  * 1.02).sum())
    
    # Breakout detection
    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98
    
    # Volume on breakout
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)
    
    # Bell shape in consolidation zone
    vp            = calculate_volume_profile(
                        df.iloc[consol_start:],
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False
    
    # Duration label
    if consol_days < 30:
        duration_label = f'{consol_days}d (weeks)'
    elif consol_days < 90:
        duration_label = f'{consol_days}d (months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (quarters)'
    else:
        duration_label = f'{consol_days}d (years)'
    
    is_valid = (
        is_sideways
        and res_touches >= 2
        and sup_touches >= 2
        and range_pct < max_range_pct
    )
    
    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : round(range_high,  2),
        'range_low'           : round(range_low,   2),
        'range_pct'           : round(range_pct,   2),
        'is_sideways'         : is_sideways,
        'norm_slope'          : round(norm_slope,  4),
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above,   2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else 'Invalid consolidation'
    }


# ── CORRECTED BREAKOUT SCORE ──────────────────
def score_breakout_v3(df):
    """
    Score breakout using fully dynamic consolidation detection
    """
    result = detect_consolidation_dynamic(df)
    scores = {}

    # No valid consolidation
    if not result.get('is_valid'):
        return 0, result

    # ── CONSOLIDATION DURATION (20 pts) ──
    days = result['consol_days']
    if days >= 250:    scores['Duration'] = 20   # 1 year+ base ✅
    elif days >= 90:   scores['Duration'] = 15   # Quarterly base
    elif days >= 30:   scores['Duration'] = 10   # Monthly base
    else:              scores['Duration'] = 5    # Short base

    # ── RANGE TIGHTNESS (15 pts) ──
    rng = result['range_pct']
    if rng < 10:       scores['Tightness'] = 15  # Very tight ✅
    elif rng < 15:     scores['Tightness'] = 12
    elif rng < 20:     scores['Tightness'] = 8
    else:              scores['Tightness'] = 4

    # ── MULTIPLE TOUCHES (15 pts) ──
    touches = min(result['resistance_touches'],
                  result['support_touches'])
    if touches >= 5:   scores['Touches'] = 15    # Well tested ✅
    elif touches >= 3: scores['Touches'] = 10
    elif touches >= 2: scores['Touches'] = 6
    else:              scores['Touches'] = 0

    # ── BELL SHAPE (15 pts) ──
    if result['has_bell']:
        scores['Bell'] = 15                       # Bell confirmed ✅
    else:
        scores['Bell'] = 0

    # ── BREAKOUT HAPPENING (20 pts) ──
    if result['is_breaking_out']:
        pct = result['pct_above']
        if pct > 3:    scores['Breakout'] = 20   # Clear breakout ✅
        elif pct > 1:  scores['Breakout'] = 15
        else:          scores['Breakout'] = 10   # Testing level
    else:
        scores['Breakout'] = 0

    # ── VOLUME CONFIRMATION (15 pts) ──
    vol = result['breakout_volume']
    if vol > 2.5:      scores['Volume'] = 15     # Strong volume ✅
    elif vol > 2.0:    scores['Volume'] = 12
    elif vol > 1.5:    scores['Volume'] = 8
    else:              scores['Volume'] = 0      # No volume confirmation

    total = sum(scores.values())
    return min(total, 100), {**scores, **result}


# ── TEST ON ALL STOCKS ────────────────────────
print("=== DYNAMIC BREAKOUT DETECTION (V3) ===\n")
print(f"{'Symbol':15} {'Score':6} {'Valid':6} {'Days':6} "
      f"{'Range%':8} {'Bell':5} {'Breaking':9} {'Vol':5}")
print("─" * 75)

for symbol in list(indicator_data.keys()):
    df    = indicator_data[symbol]
    score, detail = score_breakout_v3(df)

    valid    = detail.get('is_valid',        False)
    days     = detail.get('consol_days',     0)
    rng      = detail.get('range_pct',       0)
    bell     = detail.get('has_bell',        False)
    breaking = detail.get('is_breaking_out', False)
    vol      = detail.get('breakout_volume', 0)

    print(f"{symbol:15} {score:5}/100 "
          f"{str(valid):6} {days:6} "
          f"{rng:7.1f}% "
          f"{str(bell):5} "
          f"{str(breaking):9} "
          f"{vol:.1f}x")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=15,
                                  max_range_pct=30):
    """
    Fully dynamic consolidation detection
    No slope check on short windows — only range check
    Slope check only applied on final result
    """
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    # Initialize with last 3 days
    range_high   = high.iloc[-3:].max()
    range_low    = low.iloc[-3:].min()
    consol_start = n - 3

    # Expand backwards — ONLY range check, no slope check
    for i in range(n - 4, max(0, n - 1000), -1):

        bar_high  = high.iloc[i]
        bar_low   = low.iloc[i]

        # Tentatively expand range
        new_high  = max(range_high, bar_high)
        new_low   = min(range_low,  bar_low)
        new_range = (new_high - new_low) / new_low * 100

        # Stop if range becomes too wide
        if new_range > max_range_pct:
            break

        # Valid bar — extend consolidation
        range_high   = new_high
        range_low    = new_low
        consol_start = i

    consol_days   = n - consol_start
    current_price = close.iloc[-1]

    # Minimum duration check
    if consol_days < min_days:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': consol_days,
            'range_pct'  : round(
                (range_high - range_low) / range_low * 100, 2
            ),
            'reason'     : f'Too short: {consol_days} days'
        }

    # Final metrics
    range_pct     = (range_high - range_low) / range_low * 100

    # Slope check on FULL consolidation period only
    consol_prices = close.iloc[consol_start:].values
    x             = np.arange(len(consol_prices))
    slope         = np.polyfit(x, consol_prices, 1)[0]
    norm_slope    = abs(slope) / consol_prices.mean() * 100
    is_sideways   = norm_slope < 0.15

    # Count touches of resistance and support
    consol_df     = df.iloc[consol_start:]
    res_touches   = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches   = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    # Breakout detection
    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98

    # Volume on breakout
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)

    # Bell shape in consolidation zone
    vp            = calculate_volume_profile(
                        df.iloc[consol_start:],
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False

    # Duration label
    if consol_days < 30:
        duration_label = f'{consol_days}d (weeks)'
    elif consol_days < 90:
        duration_label = f'{consol_days}d (1-3 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (3-12 months)'
    else:
        duration_label = f'{consol_days}d (1+ year)'

    is_valid = (
        is_sideways
        and res_touches >= 2
        and sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : round(range_high,  2),
        'range_low'           : round(range_low,   2),
        'range_pct'           : round(range_pct,   2),
        'norm_slope'          : round(norm_slope,  4),
        'is_sideways'         : is_sideways,
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above,   2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'Slope:{norm_slope:.3f} '
                                f'Range:{range_pct:.1f}% '
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test
print("=== DYNAMIC BREAKOUT DETECTION (FIXED V2) ===\n")
print(f"{'Symbol':15} {'Valid':6} {'Days':6} "
      f"{'Range%':8} {'Slope':7} {'Bell':5} "
      f"{'Breaking':9} {'Vol':5}")
print("─" * 75)

for symbol in list(indicator_data.keys()):
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)

    valid    = result.get('is_valid',        False)
    days     = result.get('consol_days',     0)
    rng      = result.get('range_pct',       0)
    slope    = result.get('norm_slope',      0)
    bell     = result.get('has_bell',        False)
    breaking = result.get('is_breaking_out', False)
    vol      = result.get('breakout_volume', 0)
    reason   = result.get('reason',          '')

    status = '✅' if valid else '❌'
    print(f"{status} {symbol:15} {str(valid):6} {days:6} "
          f"{rng:7.1f}% {slope:6.3f}  "
          f"{str(bell):5} {str(breaking):9} "
          f"{vol:.1f}x  {reason}")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=15,
                                  max_range_pct=25,
                                  slope_threshold=0.13,
                                  inside_pct_required=0.85):
    """
    Improved consolidation with fuzzy range
    inside_pct_required = % of bars that must be inside range
    0.85 = 85% of bars inside = consolidation
    Allows 15% fake breakouts
    """
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    best_result = None
    best_days   = 0

    # Try different lookback windows
    # Start from min_days and keep expanding
    # Find the longest valid consolidation
    for lookback in range(min_days, min(1000, n - 5)):

        segment       = df.iloc[n - lookback - 5: n - 5]
        seg_high      = segment['High']
        seg_low       = segment['Low']
        seg_close     = segment['Close']

        seg_range_high = seg_high.max()
        seg_range_low  = seg_low.min()
        seg_range_pct  = (
            (seg_range_high - seg_range_low) /
             seg_range_low * 100
        )

        # Skip if range too wide
        if seg_range_pct > max_range_pct:
            break

        # ── FUZZY RANGE CHECK ──
        # What % of bars are inside the range
        # Allow top 7.5% and bottom 7.5% as buffer
        buffer_high = seg_range_high * 1.075
        buffer_low  = seg_range_low  * 0.925

        bars_inside = (
            (seg_high <= buffer_high) &
            (seg_low  >= buffer_low)
        ).sum()
        pct_inside  = bars_inside / len(segment)

        if pct_inside < inside_pct_required:
            break

        # ── SLOPE CHECK on highs and lows ──
        x            = np.arange(len(segment))

        high_slope   = np.polyfit(x, seg_high.values, 1)[0]
        high_norm    = high_slope / seg_high.mean() * 100

        low_slope    = np.polyfit(x, seg_low.values, 1)[0]
        low_norm     = low_slope / seg_low.mean() * 100

        close_slope  = np.polyfit(x, seg_close.values, 1)[0]
        close_norm   = abs(close_slope) / seg_close.mean() * 100

        is_sideways  = (
            abs(high_norm)  < slope_threshold and
            abs(low_norm)   < slope_threshold and
            close_norm      < slope_threshold
        )

        if not is_sideways:
            break

        # Valid consolidation found — keep expanding
        best_days   = lookback
        best_result = {
            'consol_days' : lookback,
            'range_high'  : round(seg_range_high, 2),
            'range_low'   : round(seg_range_low,  2),
            'range_pct'   : round(seg_range_pct,  2),
            'high_slope'  : round(high_norm,       4),
            'low_slope'   : round(low_norm,        4),
            'close_slope' : round(close_norm,      4),
            'pct_inside'  : round(pct_inside,      3),
            'is_sideways' : is_sideways,
        }

    # No valid consolidation found
    if best_result is None:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': 0,
            'reason'     : 'No consolidation found'
        }

    # Use the best (longest) consolidation found
    consol_days    = best_result['consol_days']
    range_high     = best_result['range_high']
    range_low      = best_result['range_low']
    current_price  = close.iloc[-1]

    # Count resistance and support touches
    consol_df      = df.iloc[n - consol_days - 5: n - 5]
    res_touches    = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches    = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    # Breakout detection — check last 5 days
    recent_high    = df['High'].tail(5).max()
    pct_above      = (current_price - range_high) / range_high * 100
    is_breaking    = (
        current_price >= range_high * 0.98 or
        recent_high   >= range_high * 1.02
    )

    # Volume on breakout
    vol_ratio      = round(df['Vol_Ratio'].tail(5).max(), 2)

    # Bell shape
    vp             = calculate_volume_profile(
                         consol_df,
                         lookback_days=consol_days,
                         bins=20
                     )
    has_bell       = vp['is_bell'] if vp else False

    # Duration label
    if consol_days < 30:
        duration_label = f'{consol_days}d (weeks)'
    elif consol_days < 90:
        duration_label = f'{consol_days}d (1-3 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (3-12 months)'
    else:
        duration_label = f'{consol_days}d (1+ year)'

    is_valid = (
        best_result['is_sideways'] and
        res_touches >= 2           and
        sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : range_high,
        'range_low'           : range_low,
        'range_pct'           : best_result['range_pct'],
        'high_slope'          : best_result['high_slope'],
        'low_slope'           : best_result['low_slope'],
        'pct_inside'          : best_result['pct_inside'],
        'is_sideways'         : best_result['is_sideways'],
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above, 2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'HiSlp:{best_result["high_slope"]:.3f} '
                                f'LoSlp:{best_result["low_slope"]:.3f} '
                                f'Inside:{best_result["pct_inside"]:.0%} '
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test on all stocks
print("=== DYNAMIC BREAKOUT V5 (FUZZY RANGE) ===\n")
print(f"{'Symbol':15} {'V':2} {'Days':5} {'Rng%':6} "
      f"{'HiSlp':7} {'LoSlp':7} {'In%':5} "
      f"{'Bell':5} {'Brk':5} {'Vol':5}")
print("─" * 78)

for symbol in list(indicator_data.keys()):
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)

    v    = '✅' if result.get('is_valid')        else '❌'
    days = result.get('consol_days',     0)
    rng  = result.get('range_pct',       0)
    hi   = result.get('high_slope',      0)
    lo   = result.get('low_slope',       0)
    ins  = result.get('pct_inside',      0)
    bell = '✅' if result.get('has_bell')         else '❌'
    brk  = '✅' if result.get('is_breaking_out')  else '❌'
    vol  = result.get('breakout_volume', 0)

    print(f"{v} {symbol:15} {days:5} {rng:5.1f}% "
          f"{hi:+6.3f} {lo:+6.3f} {ins:4.0%}  "
          f"{bell}  {brk}  {vol:.1f}x")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=15,
                                  max_range_pct=25,
                                  slope_threshold=0.13,
                                  inside_pct_required=0.85):
    """
    Find consolidation by:
    1. Start from a point in the past
    2. Define range from that segment
    3. Check if price stayed mostly inside that range
    4. Find the longest such period ending near today
    """
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    best_result = None

    # Try different consolidation end points
    # Consolidation should end in last 10 days
    for end_offset in range(0, 10):
        end_idx = n - end_offset

        # Try different consolidation lengths
        for lookback in range(min_days, min(500, end_idx - 5)):

            start_idx = end_idx - lookback
            segment   = df.iloc[start_idx:end_idx]

            if len(segment) < min_days:
                continue

            seg_high  = segment['High']
            seg_low   = segment['Low']
            seg_close = segment['Close']

            # Define range
            range_high = seg_high.quantile(0.95)  # Use 95th percentile
            range_low  = seg_low.quantile(0.05)   # Use 5th percentile
            range_pct  = (range_high - range_low) / range_low * 100

            # Skip if range too wide
            if range_pct > max_range_pct:
                break

            # Check % of bars inside range (with buffer)
            buffer     = 0.03  # 3% buffer
            bars_inside = (
                (seg_high <= range_high * (1 + buffer)) &
                (seg_low  >= range_low  * (1 - buffer))
            ).sum()
            pct_inside = bars_inside / len(segment)

            if pct_inside < inside_pct_required:
                break

            # Slope check
            x          = np.arange(len(segment))

            high_slope = np.polyfit(x, seg_high.values, 1)[0]
            high_norm  = high_slope / seg_high.mean() * 100

            low_slope  = np.polyfit(x, seg_low.values, 1)[0]
            low_norm   = low_slope / seg_low.mean() * 100

            close_slope = np.polyfit(x, seg_close.values, 1)[0]
            close_norm  = abs(close_slope) / seg_close.mean() * 100

            is_sideways = (
                abs(high_norm)  < slope_threshold and
                abs(low_norm)   < slope_threshold and
                close_norm      < slope_threshold
            )

            if not is_sideways:
                break

            # Valid — save if longest so far
            if lookback > (best_result['consol_days']
                           if best_result else 0):
                best_result = {
                    'consol_days' : lookback,
                    'end_offset'  : end_offset,
                    'range_high'  : round(range_high, 2),
                    'range_low'   : round(range_low,  2),
                    'range_pct'   : round(range_pct,  2),
                    'high_slope'  : round(high_norm,  4),
                    'low_slope'   : round(low_norm,   4),
                    'close_slope' : round(close_norm, 4),
                    'pct_inside'  : round(pct_inside, 3),
                    'is_sideways' : is_sideways,
                    'start_idx'   : start_idx,
                    'end_idx'     : end_idx,
                }

    # No valid consolidation found
    if best_result is None:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': 0,
            'reason'     : 'No consolidation found'
        }

    # Final metrics from best result
    consol_days   = best_result['consol_days']
    range_high    = best_result['range_high']
    range_low     = best_result['range_low']
    current_price = close.iloc[-1]

    # Count touches in consolidation zone
    consol_df     = df.iloc[best_result['start_idx']:
                            best_result['end_idx']]
    res_touches   = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches   = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    # Breakout — is current price above range?
    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98

    # Volume
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)

    # Bell shape
    vp            = calculate_volume_profile(
                        consol_df,
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False

    # Duration label
    if consol_days < 30:
        duration_label = f'{consol_days}d (weeks)'
    elif consol_days < 90:
        duration_label = f'{consol_days}d (1-3 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (3-12 months)'
    else:
        duration_label = f'{consol_days}d (1+ year)'

    is_valid = (
        best_result['is_sideways'] and
        res_touches >= 2           and
        sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : range_high,
        'range_low'           : range_low,
        'range_pct'           : best_result['range_pct'],
        'high_slope'          : best_result['high_slope'],
        'low_slope'           : best_result['low_slope'],
        'pct_inside'          : best_result['pct_inside'],
        'is_sideways'         : best_result['is_sideways'],
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above, 2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'HiSlp:{best_result["high_slope"]:.3f} '
                                f'LoSlp:{best_result["low_slope"]:.3f} '
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test
print("=== DYNAMIC BREAKOUT V6 ===\n")
print(f"{'Symbol':15} {'V':2} {'Days':5} {'Rng%':6} "
      f"{'HiSlp':7} {'LoSlp':7} {'In%':5} "
      f"{'Bell':5} {'Brk':5} {'Vol':5}")
print("─" * 78)

for symbol in list(indicator_data.keys()):
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)

    v    = '✅' if result.get('is_valid')        else '❌'
    days = result.get('consol_days',     0)
    rng  = result.get('range_pct',       0)
    hi   = result.get('high_slope',      0)
    lo   = result.get('low_slope',       0)
    ins  = result.get('pct_inside',      0)
    bell = '✅' if result.get('has_bell')         else '❌'
    brk  = '✅' if result.get('is_breaking_out')  else '❌'
    vol  = result.get('breakout_volume', 0)

    print(f"{v} {symbol:15} {days:5} {rng:5.1f}% "
          f"{hi:+6.3f} {lo:+6.3f} {ins:4.0%}  "
          f"{bell}  {brk}  {vol:.1f}x")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=15,
                                  max_range_pct=25,
                                  slope_threshold=0.13,
                                  inside_pct_required=0.85):
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    best_result = None

    for lookback in range(min_days, min(500, n - 5)):

        start_idx  = n - lookback
        segment    = df.iloc[start_idx:n]

        seg_high   = segment['High']
        seg_low    = segment['Low']
        seg_close  = segment['Close']

        # Range check
        range_high = seg_high.quantile(0.95)
        range_low  = seg_low.quantile(0.05)
        range_pct  = (range_high - range_low) / range_low * 100

        if range_pct > max_range_pct:
            continue

        # Inside % check
        buffer      = 0.03
        bars_inside = (
            (seg_high <= range_high * (1 + buffer)) &
            (seg_low  >= range_low  * (1 - buffer))
        ).sum()
        pct_inside  = bars_inside / len(segment)

        if pct_inside < inside_pct_required:
            continue

        # Slope check
        x           = np.arange(len(segment))
        high_slope  = np.polyfit(x, seg_high.values,  1)[0]
        high_norm   = high_slope / seg_high.mean() * 100
        low_slope   = np.polyfit(x, seg_low.values,   1)[0]
        low_norm    = low_slope  / seg_low.mean()  * 100
        close_slope = np.polyfit(x, seg_close.values, 1)[0]
        close_norm  = abs(close_slope) / seg_close.mean() * 100

        is_sideways = (
            abs(high_norm)  < slope_threshold and
            abs(low_norm)   < slope_threshold and
            close_norm      < slope_threshold
        )

        if not is_sideways:
            continue

        # ── CHECK 1: Price change during consolidation ──
        # Price must not have fallen more than 12%
        price_start  = seg_close.iloc[0]
        price_end    = seg_close.iloc[-1]
        price_change = (price_end - price_start) / price_start * 100

        if price_change < -12:
            continue   # Falling during "consolidation" = downtrend

        # ── CHECK 2: Pre-trend check ──
        pre_start    = max(0, start_idx - 60)
        pre_segment  = df.iloc[pre_start:start_idx]
        pre_close    = pre_segment['Close'].values

        if len(pre_close) > 5:
            pre_x    = np.arange(len(pre_close))
            pre_slope = np.polyfit(pre_x, pre_close, 1)[0]
            pre_norm = pre_slope / pre_close.mean() * 100

            # If strongly falling before = downtrend consolidation
            if pre_norm < -0.20:
                continue

        # ── CHECK 3: Consolidation not happening after a big drop ──
        # If price is more than 25% below its level before consolidation
        # it means price dropped significantly then went flat
        # That's not a breakout setup
        if len(pre_close) > 0:
            pre_high     = df['High'].iloc[
                max(0, start_idx-120):start_idx
            ].max()
            if pre_high > 0:
                drop_from_high = (
                    range_high - pre_high
                ) / pre_high * 100
                # Consolidation range top is 25%+ below recent high
                if drop_from_high < -25:
                    continue

        # Valid consolidation
        if lookback > (best_result['consol_days']
                       if best_result else 0):
            best_result = {
                'consol_days' : lookback,
                'range_high'  : round(range_high, 2),
                'range_low'   : round(range_low,  2),
                'range_pct'   : round(range_pct,  2),
                'high_slope'  : round(high_norm,  4),
                'low_slope'   : round(low_norm,   4),
                'close_slope' : round(close_norm, 4),
                'pct_inside'  : round(pct_inside, 3),
                'price_change': round(price_change, 2),
                'is_sideways' : True,
                'start_idx'   : start_idx,
            }

    if best_result is None:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': 0,
            'reason'     : 'No consolidation found'
        }

    consol_days   = best_result['consol_days']
    range_high    = best_result['range_high']
    range_low     = best_result['range_low']
    current_price = close.iloc[-1]

    consol_df     = df.iloc[best_result['start_idx']:n]
    res_touches   = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches   = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)

    vp            = calculate_volume_profile(
                        consol_df,
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False

    if consol_days < 30:
        duration_label = f'{consol_days}d (weeks)'
    elif consol_days < 90:
        duration_label = f'{consol_days}d (1-3 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (3-12 months)'
    else:
        duration_label = f'{consol_days}d (1+ year)'

    is_valid = (
        res_touches >= 2 and
        sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : range_high,
        'range_low'           : range_low,
        'range_pct'           : best_result['range_pct'],
        'high_slope'          : best_result['high_slope'],
        'low_slope'           : best_result['low_slope'],
        'pct_inside'          : best_result['pct_inside'],
        'price_change'        : best_result['price_change'],
        'is_sideways'         : True,
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above, 2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test
print("=== DYNAMIC BREAKOUT V10 ===\n")
print(f"{'Symbol':15} {'V':2} {'Days':5} {'Rng%':6} "
      f"{'Chg%':6} {'Bell':5} {'Brk':5} {'Vol':5}")
print("─" * 60)

for symbol in list(indicator_data.keys()):
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)

    v      = '✅' if result.get('is_valid')        else '❌'
    days   = result.get('consol_days',     0)
    rng    = result.get('range_pct',       0)
    chg    = result.get('price_change',    0)
    bell   = '✅' if result.get('has_bell')         else '❌'
    brk    = '✅' if result.get('is_breaking_out')  else '❌'
    vol    = result.get('breakout_volume', 0)
    hi     = result.get('high_slope',      0)
    lo     = result.get('low_slope',       0)

    print(f"{v} {symbol:15} {days:5} {rng:5.1f}% "
          f"{chg:+5.1f}%  {bell}  {brk}  {vol:.1f}x  "
          f"Hi:{hi:+.3f} Lo:{lo:+.3f}")
```

```python
def score_breakout_final(df):
    """
    Score breakout setup using V10 detection
    """
    result = detect_consolidation_dynamic(df)
    scores = {}

    # No consolidation found
    if not result.get('found') or not result.get('is_valid'):
        return 0, result

    # ── CONSOLIDATION DURATION (20 pts) ──
    days = result['consol_days']
    if days >= 250:    scores['Duration'] = 20
    elif days >= 90:   scores['Duration'] = 15
    elif days >= 30:   scores['Duration'] = 10
    else:              scores['Duration'] = 5

    # ── RANGE TIGHTNESS (15 pts) ──
    rng = result['range_pct']
    if rng < 10:       scores['Tightness'] = 15
    elif rng < 15:     scores['Tightness'] = 12
    elif rng < 20:     scores['Tightness'] = 8
    else:              scores['Tightness'] = 4

    # ── BELL SHAPE (15 pts) ──
    if result['has_bell']:
        scores['Bell'] = 15
    else:
        scores['Bell'] = 0

    # ── BREAKOUT STATUS (25 pts) ──
    pct = result['pct_above']
    if result['is_breaking_out'] and pct > 3:
        scores['Breakout'] = 25   # Clear breakout ✅
    elif result['is_breaking_out']:
        scores['Breakout'] = 18   # Just at level
    elif pct > -3:
        scores['Breakout'] = 12   # Very close to breakout
    elif pct > -8:
        scores['Breakout'] = 6    # Approaching
    else:
        scores['Breakout'] = 0    # Far from breakout

    # ── VOLUME (15 pts) ──
    vol = result['breakout_volume']
    if vol > 2.5:      scores['Volume'] = 15
    elif vol > 2.0:    scores['Volume'] = 12
    elif vol > 1.5:    scores['Volume'] = 8
    else:              scores['Volume'] = 0

    # ── TOUCHES (10 pts) ──
    touches = min(
        result['resistance_touches'],
        result['support_touches']
    )
    if touches >= 5:   scores['Touches'] = 10
    elif touches >= 3: scores['Touches'] = 7
    elif touches >= 2: scores['Touches'] = 4
    else:              scores['Touches'] = 0

    total = sum(scores.values())
    return min(total, 100), {**scores, **result}


# Run for all 35 stocks
print("=== FINAL BREAKOUT SCORES ===\n")
print(f"{'Symbol':15} {'Score':6} {'Days':6} "
      f"{'Rng%':6} {'Pct Above':10} {'Vol':5} {'Bell':5}")
print("─" * 65)

breakout_results = []
for _, fund_row in fundamental_df.iterrows():
    symbol = fund_row['Symbol']
    if symbol not in indicator_data:
        continue

    df             = indicator_data[symbol]
    score, detail  = score_breakout_final(df)

    breakout_results.append({
        'Symbol'          : symbol,
        'Breakout Score'  : score,
        'Consol Days'     : detail.get('consol_days', 0),
        'Range Pct'       : detail.get('range_pct',   0),
        'Pct Above'       : detail.get('pct_above',   0),
        'Breaking Out'    : detail.get('is_breaking_out', False),
        'Bell Shape'      : detail.get('has_bell',    False),
        'Volume'          : detail.get('breakout_volume', 0),
    })

    days  = detail.get('consol_days',     0)
    rng   = detail.get('range_pct',       0)
    pct   = detail.get('pct_above',       0)
    vol   = detail.get('breakout_volume', 0)
    bell  = '✅' if detail.get('has_bell') else '❌'

    print(f"{symbol:15} {score:5}/100 {days:6} "
          f"{rng:5.1f}%  {pct:+7.1f}%  {vol:.1f}x  {bell}")

breakout_df = pd.DataFrame(breakout_results).sort_values(
    'Breakout Score', ascending=False
).reset_index(drop=True)

print(f"\nTop 10 Breakout Candidates:")
for i, row in breakout_df.head(10).iterrows():
    print(f"  {i+1:2}. {row['Symbol']:15} "
          f"{row['Breakout Score']:3}/100  "
          f"{row['Consol Days']:3}d  "
          f"{row['Pct Above']:+.1f}% from breakout")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=90,
                                  max_range_pct=50,
                                  slope_threshold=0.13,
                                  inside_pct_required=0.82):
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    best_result = None

    for lookback in range(min_days, min(800, n - 5)):

        start_idx  = n - lookback
        segment    = df.iloc[start_idx:n]

        seg_high   = segment['High']
        seg_low    = segment['Low']
        seg_close  = segment['Close']

        # Range check
        range_high = seg_high.quantile(0.95)
        range_low  = seg_low.quantile(0.05)
        range_pct  = (range_high - range_low) / range_low * 100

        if range_pct > max_range_pct:
            continue

        # Inside % check
        buffer      = 0.03
        bars_inside = (
            (seg_high <= range_high * (1 + buffer)) &
            (seg_low  >= range_low  * (1 - buffer))
        ).sum()
        pct_inside  = bars_inside / len(segment)

        if pct_inside < inside_pct_required:
            continue

        # Slope check
        x           = np.arange(len(segment))
        high_slope  = np.polyfit(x, seg_high.values,  1)[0]
        high_norm   = high_slope / seg_high.mean() * 100
        low_slope   = np.polyfit(x, seg_low.values,   1)[0]
        low_norm    = low_slope  / seg_low.mean()  * 100
        close_slope = np.polyfit(x, seg_close.values, 1)[0]
        close_norm  = abs(close_slope) / seg_close.mean() * 100

        # Slight relaxation for longer periods
        slope_limit = min(slope_threshold + (lookback / 5000), 0.20)

        is_sideways = (
            abs(high_norm)  < slope_limit and
            abs(low_norm)   < slope_limit and
            close_norm      < slope_limit
        )

        if not is_sideways:
            continue

        # Price change during consolidation
        price_start  = seg_close.iloc[0]
        price_end    = seg_close.iloc[-1]
        price_change = (price_end - price_start) / price_start * 100

        # Allow more drop for longer consolidations
        max_drop = -20 if lookback < 365 else -35
        if price_change < max_drop:
            continue

        # Pre-trend check — relaxed for long consolidations
        pre_start   = max(0, start_idx - 120)
        pre_segment = df.iloc[pre_start:start_idx]
        pre_close   = pre_segment['Close'].values

        if len(pre_close) > 5:
            pre_x     = np.arange(len(pre_close))
            pre_slope = np.polyfit(pre_x, pre_close, 1)[0]
            pre_norm  = pre_slope / pre_close.mean() * 100

            # Short consolidation: strict pre-trend check
            # Long consolidation: relaxed — can start after a decline
            pre_slope_limit = -0.15 if lookback < 365 else -0.60
            if pre_norm < pre_slope_limit:
                continue

            # Drop from pre-high check
            pre_high  = pre_segment['High'].max()
            drop_pct  = (range_high - pre_high) / pre_high * 100
            max_drop_from_high = -20 if lookback < 365 else -60
            if drop_pct < max_drop_from_high:
                continue

        # Valid consolidation — save if longest
        if lookback > (best_result['consol_days']
                       if best_result else 0):
            best_result = {
                'consol_days' : lookback,
                'range_high'  : round(range_high, 2),
                'range_low'   : round(range_low,  2),
                'range_pct'   : round(range_pct,  2),
                'high_slope'  : round(high_norm,  4),
                'low_slope'   : round(low_norm,   4),
                'close_slope' : round(close_norm, 4),
                'pct_inside'  : round(pct_inside, 3),
                'price_change': round(price_change, 2),
                'is_sideways' : True,
                'start_idx'   : start_idx,
            }

    if best_result is None:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': 0,
            'reason'     : 'No consolidation found'
        }

    consol_days   = best_result['consol_days']
    range_high    = best_result['range_high']
    range_low     = best_result['range_low']
    current_price = close.iloc[-1]

    consol_df     = df.iloc[best_result['start_idx']:n]
    res_touches   = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches   = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)

    vp            = calculate_volume_profile(
                        consol_df,
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False

    if consol_days < 180:
        duration_label = f'{consol_days}d (1-6 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (6-12 months)'
    elif consol_days < 500:
        duration_label = f'{consol_days}d (1-1.5 years)'
    elif consol_days < 730:
        duration_label = f'{consol_days}d (1.5-2 years)'
    else:
        duration_label = f'{consol_days}d (2+ years)'

    is_valid = (
        res_touches >= 2 and
        sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : range_high,
        'range_low'           : range_low,
        'range_pct'           : best_result['range_pct'],
        'high_slope'          : best_result['high_slope'],
        'low_slope'           : best_result['low_slope'],
        'pct_inside'          : best_result['pct_inside'],
        'price_change'        : best_result['price_change'],
        'is_sideways'         : True,
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above, 2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test FINEORG specifically first
print("=== FINEORG CONSOLIDATION CHECK ===")
result = detect_consolidation_dynamic(indicator_data['FINEORG'])
print(f"Found       : {result['found']}")
print(f"Valid       : {result['is_valid']}")
print(f"Days        : {result['consol_days']}")
print(f"Range       : ₹{result['range_low']} — ₹{result['range_high']}")
print(f"Range %     : {result['range_pct']}%")
print(f"Duration    : {result['duration_label']}")
print(f"Pct above   : {result['pct_above']}%")
print(f"Bell shape  : {result['has_bell']}")
```

```python
def check_volume_concentration(df, lookback_days=600, bins=20):
    """
    Check if volume is concentrated in middle of range
    True consolidation = bell shape = high middle concentration
    False consolidation = flat distribution = low concentration
    """
    recent     = df.tail(lookback_days).copy()
    price_min  = recent['Low'].min()
    price_max  = recent['High'].max()

    if price_max <= price_min:
        return None

    price_bins       = np.linspace(price_min, price_max, bins + 1)
    volume_at_price  = np.zeros(bins)

    for _, row in recent.iterrows():
        candle_low   = row['Low']
        candle_high  = row['High']
        candle_vol   = row['Volume']
        candle_range = candle_high - candle_low

        if candle_range == 0:
            continue

        for i in range(bins):
            bin_low  = price_bins[i]
            bin_high = price_bins[i + 1]

            overlap_low  = max(candle_low,  bin_low)
            overlap_high = min(candle_high, bin_high)

            if overlap_high > overlap_low:
                overlap_pct         = (
                    overlap_high - overlap_low
                ) / candle_range
                volume_at_price[i] += candle_vol * overlap_pct

    total_volume = volume_at_price.sum()
    if total_volume == 0:
        return None

    # Middle 50% of range
    mid_start = bins // 4       # 25% from bottom
    mid_end   = 3 * bins // 4   # 75% from bottom

    middle_volume = volume_at_price[mid_start:mid_end].sum()
    concentration = middle_volume / total_volume * 100

    # POC index
    poc_idx   = np.argmax(volume_at_price)
    poc_price = (price_bins[poc_idx] + price_bins[poc_idx + 1]) / 2

    # Is POC in middle of range? (not at edges)
    poc_position = poc_idx / bins * 100  # % from bottom
    poc_in_middle = 20 < poc_position < 80

    # Bell shape score
    # Compare middle volume to edge volume
    edge_vol    = (
        volume_at_price[:3].mean() +
        volume_at_price[-3:].mean()
    ) / 2
    peak_vol    = volume_at_price.max()
    bell_ratio  = peak_vol / (edge_vol + 1)

    return {
        'concentration'  : round(concentration, 1),
        'poc_price'      : round(poc_price, 2),
        'poc_position'   : round(poc_position, 1),
        'poc_in_middle'  : poc_in_middle,
        'bell_ratio'     : round(bell_ratio, 2),
        'is_bell'        : bell_ratio > 2.5 and poc_in_middle,
    }


# Compare TCS vs FINEORG
for symbol in ['TCS', 'FINEORG', 'IPCALAB',
               'SUNPHARMA', 'TITAN', 'PERSISTENT']:
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)
    days   = result.get('consol_days', 0)

    if days > 0:
        vc = check_volume_concentration(df, lookback_days=days)
        if vc:
            print(f"{symbol:15}"
                  f"  Days:{days:4}"
                  f"  MidConc:{vc['concentration']:5.1f}%"
                  f"  BellRatio:{vc['bell_ratio']:5.2f}"
                  f"  POC@{vc['poc_position']:4.1f}%"
                  f"  Bell:{vc['is_bell']}"
                  f"  {'✅ TRUE CONSOL' if vc['is_bell'] else '❌ FALSE CONSOL'}")
```

```python
# Check price path shape
# True consolidation = price oscillates around middle
# False = price made a mountain (up then down)

def check_price_path(df, lookback_days):
    """
    Distinguish true consolidation from mountain/valley shape
    """
    segment    = df.tail(lookback_days)
    close      = segment['Close'].values
    n          = len(close)
    mid        = n // 2

    # Split into first half and second half
    first_half  = close[:mid]
    second_half = close[mid:]

    first_mean  = first_half.mean()
    second_mean = second_half.mean()

    # For true consolidation:
    # First half mean ≈ Second half mean
    # Price stayed at same level throughout

    # For mountain shape (TCS):
    # First half mean > Second half mean (went up then down)
    # OR Second half mean > First half mean (went down then up)

    half_diff   = abs(first_mean - second_mean) / close.mean() * 100

    # Also check: did price make a big peak in the middle?
    first_max   = first_half.max()
    last_max    = second_half.max()
    overall_max = close.max()

    # Peak in middle = mountain shape
    peak_position = np.argmax(close) / n * 100

    is_mountain = (
        peak_position > 20 and
        peak_position < 80 and
        overall_max > close[0]  * 1.30 and
        overall_max > close[-1] * 1.30
    )

    return {
        'half_diff'     : round(half_diff, 2),
        'first_mean'    : round(first_mean, 2),
        'second_mean'   : round(second_mean, 2),
        'peak_position' : round(peak_position, 1),
        'is_mountain'   : is_mountain,
        'is_true_consol': not is_mountain and half_diff < 15,
    }


# Test on key stocks
print("=== PRICE PATH ANALYSIS ===\n")
for symbol in ['TCS', 'FINEORG', 'IPCALAB',
               'SUNPHARMA', 'TITAN', 'PERSISTENT',
               'VINATIORGA', 'RELIANCE', 'HAVELLS']:
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)
    days   = result.get('consol_days', 0)

    if days > 0:
        pp = check_price_path(df, days)
        vc = check_volume_concentration(df, lookback_days=days)

        status = '✅' if pp['is_true_consol'] else '❌'
        print(f"{status} {symbol:15}"
              f"  Days:{days:4}"
              f"  HalfDiff:{pp['half_diff']:5.1f}%"
              f"  Peak@{pp['peak_position']:5.1f}%"
              f"  Mountain:{str(pp['is_mountain']):5}"
              f"  TrueConsol:{pp['is_true_consol']}")
```

```python
# Test improved mountain detection
def check_mountain_improved(df, lookback, range_high, range_low):
    segment      = df.tail(lookback)
    close_arr    = segment['Close'].values
    peak_idx     = np.argmax(close_arr)
    peak_pos     = peak_idx / lookback * 100
    peak_price   = close_arr.max()
    price_start  = close_arr[0]
    price_end    = close_arr[-1]

    rng          = range_high - range_low
    if rng == 0:
        return False

    start_vs_range = (price_start - range_low) / rng * 100
    end_vs_range   = (price_end   - range_low) / rng * 100
    peak_vs_range  = (peak_price  - range_low) / rng * 100

    is_mountain = (
        peak_pos       > 15 and
        peak_pos       < 85 and
        start_vs_range < 30 and
        end_vs_range   < 30 and
        peak_vs_range  > 85
    )

    return is_mountain, {
        'peak_pos'       : round(peak_pos,       1),
        'start_vs_range' : round(start_vs_range, 1),
        'end_vs_range'   : round(end_vs_range,   1),
        'peak_vs_range'  : round(peak_vs_range,  1),
        'is_mountain'    : is_mountain,
    }


# Test on key stocks
print("=== IMPROVED MOUNTAIN DETECTION ===\n")
for symbol in ['TCS', 'FINEORG', 'PERSISTENT',
               'HAVELLS', 'IPCALAB', 'SUNPHARMA',
               'TITAN', 'VINATIORGA', 'MUTHOOTFIN']:
    df     = indicator_data[symbol]
    n      = len(df)

    # Get range for 400 day lookback
    lookback   = min(400, n - 5)
    segment    = df.tail(lookback)
    range_high = segment['High'].quantile(0.95)
    range_low  = segment['Low'].quantile(0.05)

    result, detail = check_mountain_improved(
        df, lookback, range_high, range_low
    )

    status = '❌ MOUNTAIN' if result else '✅ CONSOL'
    print(f"{status}  {symbol:15}"
          f"  Peak@{detail['peak_pos']:5.1f}%"
          f"  Start:{detail['start_vs_range']:5.1f}%"
          f"  End:{detail['end_vs_range']:5.1f}%"
          f"  PeakVsRange:{detail['peak_vs_range']:5.1f}%")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=90,
                                  max_range_pct=50,
                                  slope_threshold=0.13,
                                  inside_pct_required=0.82):
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    best_result = None

    for lookback in range(min_days, min(800, n - 5)):

        start_idx  = n - lookback
        segment    = df.iloc[start_idx:n]
        seg_high   = segment['High']
        seg_low    = segment['Low']
        seg_close  = segment['Close']

        # Range
        range_high = seg_high.quantile(0.95)
        range_low  = seg_low.quantile(0.05)
        range_pct  = (range_high - range_low) / range_low * 100
        rng        = range_high - range_low

        if range_pct > max_range_pct or rng == 0:
            continue

        # Inside % check
        buffer      = 0.03
        bars_inside = (
            (seg_high <= range_high * (1 + buffer)) &
            (seg_low  >= range_low  * (1 - buffer))
        ).sum()
        pct_inside  = bars_inside / len(segment)

        if pct_inside < inside_pct_required:
            continue

        # Slope check
        x           = np.arange(len(segment))
        high_slope  = np.polyfit(x, seg_high.values,  1)[0]
        high_norm   = high_slope / seg_high.mean() * 100
        low_slope   = np.polyfit(x, seg_low.values,   1)[0]
        low_norm    = low_slope  / seg_low.mean()  * 100
        close_slope = np.polyfit(x, seg_close.values, 1)[0]
        close_norm  = abs(close_slope) / seg_close.mean() * 100

        slope_limit = min(slope_threshold + (lookback / 5000), 0.20)

        is_sideways = (
            abs(high_norm)  < slope_limit and
            abs(low_norm)   < slope_limit and
            close_norm      < slope_limit
        )

        if not is_sideways:
            continue

        # Price positions within range
        price_start    = seg_close.iloc[0]
        price_end      = seg_close.iloc[-1]
        price_change   = (price_end - price_start) / price_start * 100

        start_vs_range = (price_start - range_low) / rng * 100
        end_vs_range   = (price_end   - range_low) / rng * 100

        # Rule 1: End price must be inside range
        # Not below support
        if end_vs_range < -10:
            continue

        # Rule 2: Price change absolute limit
        if price_change < -30:
            continue

        # Rule 3: Mountain shape detection
        close_arr    = seg_close.values
        peak_idx     = np.argmax(close_arr)
        peak_pos     = peak_idx / lookback * 100
        peak_price   = close_arr.max()
        peak_vs_range = (peak_price - range_low) / rng * 100

        is_mountain  = (
            peak_pos       > 20  and
            peak_pos       < 80  and
            start_vs_range < 35  and
            end_vs_range   < 35  and
            peak_vs_range  > 90
        )

        if is_mountain:
            continue

        # Pre-trend check
        pre_start   = max(0, start_idx - 120)
        pre_segment = df.iloc[pre_start:start_idx]
        pre_close   = pre_segment['Close'].values

        if len(pre_close) > 5:
            pre_x     = np.arange(len(pre_close))
            pre_slope = np.polyfit(pre_x, pre_close, 1)[0]
            pre_norm  = pre_slope / pre_close.mean() * 100

            pre_limit = -0.15 if lookback < 365 else -0.60
            if pre_norm < pre_limit:
                continue

            pre_high  = pre_segment['High'].max()
            drop_pct  = (range_high - pre_high) / pre_high * 100
            drop_limit = -20 if lookback < 365 else -60
            if drop_pct < drop_limit:
                continue

        # Valid
        if lookback > (best_result['consol_days']
                       if best_result else 0):
            best_result = {
                'consol_days'  : lookback,
                'range_high'   : round(range_high,   2),
                'range_low'    : round(range_low,    2),
                'range_pct'    : round(range_pct,    2),
                'high_slope'   : round(high_norm,    4),
                'low_slope'    : round(low_norm,     4),
                'close_slope'  : round(close_norm,   4),
                'pct_inside'   : round(pct_inside,   3),
                'price_change' : round(price_change, 2),
                'end_vs_range' : round(end_vs_range, 1),
                'is_sideways'  : True,
                'start_idx'    : start_idx,
            }

    if best_result is None:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': 0,
            'reason'     : 'No consolidation found'
        }

    consol_days   = best_result['consol_days']
    range_high    = best_result['range_high']
    range_low     = best_result['range_low']
    current_price = close.iloc[-1]

    consol_df     = df.iloc[best_result['start_idx']:n]
    res_touches   = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches   = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)

    vp            = calculate_volume_profile(
                        consol_df,
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False

    if consol_days < 180:
        duration_label = f'{consol_days}d (1-6 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (6-12 months)'
    elif consol_days < 500:
        duration_label = f'{consol_days}d (1-1.5 years)'
    elif consol_days < 730:
        duration_label = f'{consol_days}d (1.5-2 years)'
    else:
        duration_label = f'{consol_days}d (2+ years)'

    is_valid = (
        res_touches >= 2 and
        sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : range_high,
        'range_low'           : range_low,
        'range_pct'           : best_result['range_pct'],
        'high_slope'          : best_result['high_slope'],
        'low_slope'           : best_result['low_slope'],
        'pct_inside'          : best_result['pct_inside'],
        'price_change'        : best_result['price_change'],
        'end_vs_range'        : best_result['end_vs_range'],
        'is_sideways'         : True,
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above, 2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test key stocks
print("=== FINAL CONSOLIDATION TEST ===\n")
for symbol in ['TCS', 'FINEORG', 'PERSISTENT', 'HAVELLS',
               'IPCALAB', 'SUNPHARMA', 'TITAN', 'VINATIORGA',
               'MUTHOOTFIN', 'DHANUKA', 'BEL', 'RELIANCE']:
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)
    days   = result.get('consol_days', 0)
    valid  = result.get('is_valid',    False)

    status = '✅' if valid and days > 0 else '❌'

    if days > 0:
        print(f"{status} {symbol:15}"
              f" {days:4}d"
              f" Range:{result.get('range_pct'):5.1f}%"
              f" EndVsRng:{result.get('end_vs_range'):+6.1f}%"
              f" PriceChg:{result.get('price_change'):+6.1f}%"
              f" {result.get('duration_label')}")
    else:
        print(f"{status} {symbol:15} Not consolidating")
```

```python
def detect_consolidation_dynamic(df,
                                  min_days=90,
                                  max_range_pct=50,
                                  max_total_drift=10,    # % over entire period
                                  inside_pct_required=0.82):
    close  = df['Close']
    high   = df['High']
    low    = df['Low']
    n      = len(df)

    if n < min_days + 5:
        return {'found': False, 'is_valid': False,
                'reason': 'Insufficient data'}

    best_result = None

    for lookback in range(min_days, min(800, n - 5)):

        start_idx  = n - lookback
        segment    = df.iloc[start_idx:n]
        seg_high   = segment['High']
        seg_low    = segment['Low']
        seg_close  = segment['Close']

        # Range
        range_high = seg_high.quantile(0.95)
        range_low  = seg_low.quantile(0.05)
        range_pct  = (range_high - range_low) / range_low * 100
        rng        = range_high - range_low

        if range_pct > max_range_pct or rng == 0:
            continue

        # Inside % check
        buffer      = 0.03
        bars_inside = (
            (seg_high <= range_high * (1 + buffer)) &
            (seg_low  >= range_low  * (1 - buffer))
        ).sum()
        pct_inside  = bars_inside / len(segment)

        if pct_inside < inside_pct_required:
            continue

        # Total drift check — replaces per-day slope
        x            = np.arange(len(segment))
        hi_slope     = np.polyfit(x, seg_high.values,  1)[0]
        hi_norm      = hi_slope  / seg_high.mean()  * 100
        lo_slope     = np.polyfit(x, seg_low.values,   1)[0]
        lo_norm      = lo_slope  / seg_low.mean()   * 100
        cl_slope     = np.polyfit(x, seg_close.values, 1)[0]
        cl_norm      = abs(cl_slope) / seg_close.mean() * 100

        # Total drift = slope × days
        hi_drift     = abs(hi_norm) * lookback
        lo_drift     = abs(lo_norm) * lookback
        cl_drift     = cl_norm      * lookback

        is_sideways  = (
            hi_drift < max_total_drift and
            lo_drift < max_total_drift and
            cl_drift < max_total_drift
        )

        if not is_sideways:
            continue

        # Price change ±25%
        price_start  = seg_close.iloc[0]
        price_end    = seg_close.iloc[-1]
        price_change = (price_end - price_start) / price_start * 100

        if abs(price_change) > 25:
            continue

        # End price inside range
        end_vs_range = (price_end - range_low) / rng * 100
        if end_vs_range < -10:
            continue

        # Mountain check
        close_arr     = seg_close.values
        peak_idx      = np.argmax(close_arr)
        peak_pos      = peak_idx / lookback * 100
        peak_price    = close_arr.max()
        peak_vs_range = (peak_price  - range_low) / rng * 100
        start_vs_rng  = (price_start - range_low) / rng * 100

        is_mountain = (
            peak_pos      > 20  and
            peak_pos      < 80  and
            start_vs_rng  < 35  and
            end_vs_range  < 35  and
            peak_vs_range > 90
        )

        if is_mountain:
            continue

        # Pre-trend check
        pre_start   = max(0, start_idx - 120)
        pre_segment = df.iloc[pre_start:start_idx]
        pre_close   = pre_segment['Close'].values

        if len(pre_close) > 5:
            pre_x     = np.arange(len(pre_close))
            pre_slope = np.polyfit(pre_x, pre_close, 1)[0]
            pre_norm  = pre_slope / pre_close.mean() * 100

            pre_limit = -0.15 if lookback < 365 else -0.60
            if pre_norm < pre_limit:
                continue

            pre_high   = pre_segment['High'].max()
            drop_pct   = (range_high - pre_high) / pre_high * 100
            drop_limit = -20 if lookback < 365 else -60
            if drop_pct < drop_limit:
                continue

        # Valid
        if lookback > (best_result['consol_days']
                       if best_result else 0):
            best_result = {
                'consol_days'  : lookback,
                'range_high'   : round(range_high,   2),
                'range_low'    : round(range_low,    2),
                'range_pct'    : round(range_pct,    2),
                'hi_drift'     : round(hi_drift,     2),
                'lo_drift'     : round(lo_drift,     2),
                'cl_drift'     : round(cl_drift,     2),
                'pct_inside'   : round(pct_inside,   3),
                'price_change' : round(price_change, 2),
                'end_vs_range' : round(end_vs_range, 1),
                'is_sideways'  : True,
                'start_idx'    : start_idx,
            }

    if best_result is None:
        return {
            'found'      : False,
            'is_valid'   : False,
            'consol_days': 0,
            'reason'     : 'No consolidation found'
        }

    consol_days   = best_result['consol_days']
    range_high    = best_result['range_high']
    range_low     = best_result['range_low']
    current_price = close.iloc[-1]

    consol_df     = df.iloc[best_result['start_idx']:n]
    res_touches   = int(
        (consol_df['High'] >= range_high * 0.98).sum()
    )
    sup_touches   = int(
        (consol_df['Low']  <= range_low  * 1.02).sum()
    )

    pct_above     = (current_price - range_high) / range_high * 100
    is_breaking   = current_price >= range_high * 0.98
    vol_ratio     = round(df['Vol_Ratio'].tail(5).max(), 2)

    vp            = calculate_volume_profile(
                        consol_df,
                        lookback_days=consol_days,
                        bins=20
                    )
    has_bell      = vp['is_bell'] if vp else False

    if consol_days < 180:
        duration_label = f'{consol_days}d (1-6 months)'
    elif consol_days < 365:
        duration_label = f'{consol_days}d (6-12 months)'
    elif consol_days < 500:
        duration_label = f'{consol_days}d (1-1.5 years)'
    elif consol_days < 730:
        duration_label = f'{consol_days}d (1.5-2 years)'
    else:
        duration_label = f'{consol_days}d (2+ years)'

    is_valid = (
        res_touches >= 2 and
        sup_touches >= 2
    )

    return {
        'found'               : True,
        'is_valid'            : is_valid,
        'consol_days'         : consol_days,
        'duration_label'      : duration_label,
        'range_high'          : range_high,
        'range_low'           : range_low,
        'range_pct'           : best_result['range_pct'],
        'hi_drift'            : best_result['hi_drift'],
        'lo_drift'            : best_result['lo_drift'],
        'cl_drift'            : best_result['cl_drift'],
        'pct_inside'          : best_result['pct_inside'],
        'price_change'        : best_result['price_change'],
        'end_vs_range'        : best_result['end_vs_range'],
        'is_sideways'         : True,
        'resistance_touches'  : res_touches,
        'support_touches'     : sup_touches,
        'is_breaking_out'     : is_breaking,
        'pct_above'           : round(pct_above, 2),
        'breakout_volume'     : vol_ratio,
        'has_bell'            : has_bell,
        'current_price'       : round(current_price, 2),
        'reason'              : 'Valid' if is_valid else
                                f'R:{res_touches} S:{sup_touches}'
    }


# Test all key stocks
print("=== CONSOLIDATION TEST (TOTAL DRIFT) ===\n")
for symbol in ['TCS', 'FINEORG', 'PERSISTENT', 'HAVELLS',
               'IPCALAB', 'SUNPHARMA', 'TITAN', 'VINATIORGA',
               'MUTHOOTFIN', 'DHANUKA', 'BEL', 'RELIANCE',
               'DIVISLAB', 'CHOLAFIN', 'GRANULES', 'HAL',
               'GARFIBRES', 'SUPRAJIT', 'PARAS']:
    df     = indicator_data[symbol]
    result = detect_consolidation_dynamic(df)
    days   = result.get('consol_days', 0)
    valid  = result.get('is_valid',    False)
    status = '✅' if valid and days > 0 else '❌'

    if days > 0:
        print(f"{status} {symbol:15}"
              f" {days:4}d"
              f" Rng:{result.get('range_pct'):5.1f}%"
              f" Drift:{result.get('cl_drift'):5.1f}%"
              f" PrChg:{result.get('price_change'):+6.1f}%"
              f" {result.get('duration_label')}")
    else:
        print(f"{status} {symbol:15} Not consolidating")
```

```python
def get_consolidation_info(df):
    """
    Returns consolidation facts for stocks with 300+ day base
    """
    result = detect_consolidation_dynamic(df, min_days=300)

    if not result.get('found') or not result.get('is_valid'):
        return {
            'consolidating'     : False,
            'consol_days'       : 0,
            'duration_label'    : 'No consolidation',
            'range_high'        : None,
            'range_low'         : None,
            'range_pct'         : None,
            'pct_to_breakout'   : None,
            'is_breaking_out'   : False,
            'breakout_volume'   : None,
            'has_bell'          : False,
            'resistance_touches': 0,
            'support_touches'   : 0,
            'current_price'     : df['Close'].iloc[-1],
        }

    return {
        'consolidating'     : True,
        'consol_days'       : result['consol_days'],
        'duration_label'    : result['duration_label'],
        'range_high'        : result['range_high'],
        'range_low'         : result['range_low'],
        'range_pct'         : result['range_pct'],
        'pct_to_breakout'   : result['pct_above'],
        'is_breaking_out'   : result['is_breaking_out'],
        'breakout_volume'   : result['breakout_volume'],
        'has_bell'          : result['has_bell'],
        'resistance_touches': result['resistance_touches'],
        'support_touches'   : result['support_touches'],
        'current_price'     : result['current_price'],
    }


# Run final report — 300+ days only
print("=== CONSOLIDATION REPORT (300+ DAYS ONLY) ===\n")
print(f"{'Symbol':15} {'Fund':6} {'Days':6} "
      f"{'Range%':7} {'To Break':9} "
      f"{'Vol':5} {'Bell':5} {'Duration'}")
print("─" * 80)

consol_info_list = []

for _, fund_row in fundamental_df.iterrows():
    symbol = fund_row['Symbol']
    if symbol not in indicator_data:
        continue

    df   = indicator_data[symbol]
    info = get_consolidation_info(df)

    if info['consolidating']:
        consol_info_list.append({
            'Symbol'            : symbol,
            'Fund Score'        : fund_row['Final Score'],
            'Sector'            : fund_row['Sector'],
            **info
        })

        bell = '✅' if info['has_bell']        else '❌'
        brk  = '🚀' if info['is_breaking_out'] else ''

        print(f"✅ {symbol:15}"
              f" {fund_row['Final Score']:5.1f}"
              f" {info['consol_days']:6}d"
              f" {info['range_pct']:6.1f}%"
              f"  {info['pct_to_breakout']:+7.1f}%"
              f"  {info['breakout_volume']:.1f}x"
              f"  {bell}"
              f"  {info['duration_label']} {brk}")
    else:
        print(f"❌ {symbol:15} Not in long consolidation")

consol_df = pd.DataFrame(consol_info_list).sort_values(
    'consol_days', ascending=False
).reset_index(drop=True)

consol_df.to_csv('../data/consolidation_info.csv', index=False)

print(f"\n{'='*60}")
print(f"Stocks in 300+ day consolidation: {len(consol_info_list)}/35")
print(f"\n=== RANKED BY DURATION ===")
for i, row in consol_df.iterrows():
    print(f"  {i+1:2}. {row['Symbol']:15}"
          f" Fund:{row['Fund Score']:5.1f}"
          f"  {row['consol_days']:4}d"
          f"  Range:{row['range_pct']:.1f}%"
          f"  {row['pct_to_breakout']:+.1f}% to breakout"
          f"  {row['duration_label']}")
```

```python
# ============================================
# FINAL COMBINED TECHNICAL REPORT
# ============================================

def generate_technical_report(symbol, fund_score, sector):
    """
    Combine all technical analysis into one report per stock
    """
    if symbol not in indicator_data:
        return None

    df      = indicator_data[symbol]
    latest  = df.iloc[-1]

    # ── BASIC INDICATORS ──
    close   = latest['Close']
    rsi     = round(latest['RSI'],       2)
    adx     = round(latest['ADX'],       2)
    macd    = round(latest['MACD'],      2)
    signal  = round(latest['Signal'],    2)
    hist    = round(latest['MACD_Hist'], 2)
    vol_r   = round(latest['Vol_Ratio'], 2)
    ema20   = round(latest['EMA20'],     2)
    ema50   = round(latest['EMA50'],     2)
    ema200  = round(latest['EMA200'],    2)
    di_plus = round(latest['DI_Plus'],   2)
    di_minus= round(latest['DI_Minus'],  2)

    # ── SETUP SCORES ──
    vp              = calculate_volume_profile(df)
    mom_score, _    = score_momentum(df, vp)
    rev_score, _    = score_reversal(df, vp)
    consol_info     = get_consolidation_info(df)

    # ── BEST TECHNICAL SETUP ──
    if mom_score >= rev_score and mom_score >= 50:
        best_setup  = 'Momentum'
        tech_score  = mom_score
    elif rev_score >= mom_score and rev_score >= 50:
        best_setup  = 'Reversal'
        tech_score  = rev_score
    else:
        best_setup  = 'Watching'
        tech_score  = max(mom_score, rev_score)

    # ── CONSOLIDATION STATUS ──
    in_consol       = consol_info['consolidating']
    consol_days     = consol_info['consol_days']
    consol_label    = consol_info['duration_label']
    pct_to_breakout = consol_info['pct_to_breakout'] or 0
    consol_volume   = consol_info['breakout_volume']  or 0

    # ── TIER CLASSIFICATION ──
    if fund_score >= 60 and tech_score >= 65:
        if best_setup == 'Momentum':
            tier  = 'TIER 1 — BUY NOW (Momentum)'
        else:
            tier  = 'TIER 1 — BUY NOW (Reversal)'
    elif fund_score >= 60 and in_consol and pct_to_breakout > -5:
        tier      = 'TIER 1 — BREAKOUT IMMINENT'
    elif fund_score >= 60 and tech_score >= 40:
        tier      = 'TIER 2 — WATCHLIST'
    elif fund_score >= 60 and in_consol:
        tier      = 'TIER 2 — BASE BUILDING'
    else:
        tier      = 'TIER 3 — WAITING'

    # ── KEY LEVELS ──
    if vp:
        poc         = vp['poc_price']
        vah         = vp['vah']
        val         = vp['val']
    else:
        poc         = None
        vah         = None
        val         = None

    # Entry, stop, target
    if best_setup == 'Momentum':
        entry       = close
        stop        = round(ema50 * 0.97, 2)
        target      = round(close * 1.20, 2)
    elif best_setup == 'Reversal':
        entry       = close
        stop        = round(df['Low'].tail(20).min() * 0.97, 2)
        target      = round(ema200 * 1.05, 2)
    elif in_consol and consol_info['range_high']:
        entry       = round(consol_info['range_high'] * 1.02, 2)
        stop        = round(consol_info['range_low']  * 0.97, 2)
        target      = round(consol_info['range_high'] +
                      (consol_info['range_high'] -
                       consol_info['range_low']) * 0.75, 2)
    else:
        entry       = None
        stop        = None
        target      = None

    return {
        'Symbol'            : symbol,
        'Sector'            : sector,
        'Fund Score'        : fund_score,
        'Current Price'     : round(close, 2),

        # Indicators
        'RSI'               : rsi,
        'ADX'               : adx,
        'MACD Hist'         : hist,
        'Vol Ratio'         : vol_r,
        'EMA20'             : ema20,
        'EMA50'             : ema50,
        'EMA200'            : ema200,
        'DI+'               : di_plus,
        'DI-'               : di_minus,

        # Setup
        'Momentum Score'    : mom_score,
        'Reversal Score'    : rev_score,
        'Best Setup'        : best_setup,
        'Tech Score'        : tech_score,
        'Tier'              : tier,

        # Consolidation
        'In Consolidation'  : in_consol,
        'Consol Days'       : consol_days,
        'Consol Label'      : consol_label,
        'Pct to Breakout'   : round(pct_to_breakout, 2),
        'Consol Volume'     : consol_volume,

        # Levels
        'Entry'             : entry,
        'Stop Loss'         : stop,
        'Target'            : target,
        'POC'               : poc,
        'VAH'               : vah,
        'VAL'               : val,
    }


# Run for all 35 stocks
print("Generating technical report for all 35 stocks...")
print("This may take 2-3 minutes...\n")

tech_reports = []

for _, fund_row in fundamental_df.iterrows():
    symbol     = fund_row['Symbol']
    fund_score = fund_row['Final Score']
    sector     = fund_row['Sector']

    report = generate_technical_report(symbol, fund_score, sector)
    if report:
        tech_reports.append(report)
        print(f"✓ {symbol:15} {report['Tier']}")

tech_report_df = pd.DataFrame(tech_reports)
tech_report_df.to_csv('../data/technical_report.csv', index=False)

print(f"\nTotal stocks processed: {len(tech_reports)}")
print(f"Saved to technical_report.csv ✅")
```

```python
# ============================================
# PREPARE COMPLETE DATA FOR DASHBOARD
# ============================================

import json
import math

def safe_int(val, default=0):
    try:
        if val is None or (isinstance(val, float) and math.isnan(val)):
            return default
        return int(float(val))
    except:
        return default

def safe_float(val, default=0.0):
    try:
        if val is None or (isinstance(val, float) and math.isnan(val)):
            return default
        return float(val)
    except:
        return default

def safe_round(val, decimals=2, default=0.0):
    try:
        if val is None or (isinstance(val, float) and math.isnan(val)):
            return default
        return round(float(val), decimals)
    except:
        return default

# Load all data
fund_df    = pd.read_csv('../data/fundamental_scores.csv')
metrics_df = pd.read_csv('../data/fundamental_metrics.csv')
tech_df    = pd.read_csv('../data/technical_report.csv')
consol_df  = pd.read_csv('../data/consolidation_info.csv')

# ── CORRECTED RELATIVE SECTOR SCORE (1-10) ──
fund_df['Sector Score'] = 0.0

for sector in fund_df['Sector'].unique():
    sector_mask = fund_df['Sector'] == sector
    sector_df   = fund_df[sector_mask]
    max_score   = sector_df['Final Score'].max()

    # Simple ratio: stock/best × 10
    fund_df.loc[sector_mask, 'Sector Score'] = (
        sector_df['Final Score'] / max_score * 10
    ).round(1)

# Show corrected scores
print("=== CORRECTED SECTOR RELATIVE SCORES ===")
for sector in sorted(fund_df['Sector'].unique()):
    sector_stocks = fund_df[fund_df['Sector'] == sector][
        ['Symbol', 'Final Score', 'Sector Score']
    ].sort_values('Sector Score', ascending=False)
    print(f"\n{sector}:")
    for _, r in sector_stocks.iterrows():
        bar = '█' * int(r['Sector Score'])
        print(f"  {r['Symbol']:15} "
              f"Fund:{r['Final Score']:5.1f}  "
              f"Sector:{r['Sector Score']:4.1f}/10  "
              f"{bar}")
        
# ── MERGE ALL DATA ──
combined = tech_df.merge(
    fund_df[['Symbol', 'Sector Score', 'Final Score',
             'Historical Score', 'Peer Score', 'Quality Score']],
    on='Symbol', how='left'
)

combined = combined.merge(
    consol_df[['Symbol', 'consol_days', 'range_pct',
               'range_high', 'range_low',
               'pct_to_breakout', 'breakout_volume']].rename(columns={
        'consol_days'     : 'Consol Days Full',
        'range_pct'       : 'Consol Range',
        'range_high'      : 'Consol High',
        'range_low'       : 'Consol Low',
        'pct_to_breakout' : 'Pct to Breakout',
        'breakout_volume' : 'Breakout Vol',
    }),
    on='Symbol', how='left'
)

# ── BUILD STOCK CARDS ──
stock_cards = []

for _, row in combined.iterrows():
    symbol = row['Symbol']

    m = metrics_df[metrics_df['Symbol'] == symbol]
    if len(m) == 0:
        continue
    m = m.iloc[0]

    fund_details = {
        'revenue_growth'   : safe_round(m.get('Revenue Growth 3Y'), 1),
        'profit_growth'    : safe_round(m.get('Profit Growth 3Y'), 1),
        'roe'              : safe_round(m.get('Final ROE'), 1),
        'roce'             : safe_round(m.get('ROCE'), 1),
        'debt_equity'      : safe_round(m.get('Debt to Equity'), 2),
        'promoter_holding' : safe_round(m.get('Promoter Holding'), 1),
        'promoter_change'  : safe_round(m.get('Promoter Change 4Q'), 2),
        'cf_consistency'   : safe_round(m.get('CF Consistency'), 1),
        'pat_margin'       : safe_round(m.get('PAT Margin'), 1),
        'sales_ttm'        : safe_round(m.get('Sales TTM'), 0),
        'profit_ttm'       : safe_round(m.get('Profit TTM'), 0),
        'pe_ratio'         : safe_round(m.get('PE Ratio'), 1),
    }

    tech_details = {
        'rsi'            : safe_round(row.get('RSI'),           1),
        'adx'            : safe_round(row.get('ADX'),           1),
        'macd_hist'      : safe_round(row.get('MACD Hist'),     2),
        'vol_ratio'      : safe_round(row.get('Vol Ratio'),     2),
        'ema20'          : safe_round(row.get('EMA20'),         2),
        'ema50'          : safe_round(row.get('EMA50'),         2),
        'ema200'         : safe_round(row.get('EMA200'),        2),
        'di_plus'        : safe_round(row.get('DI+'),           1),
        'di_minus'       : safe_round(row.get('DI-'),           1),
        'momentum_score' : safe_int(row.get('Momentum Score')),
        'reversal_score' : safe_int(row.get('Reversal Score')),
    }

    in_consol = bool(row.get('In Consolidation', False))
    consol_details = {
        'in_consolidation': in_consol,
        'days'            : safe_int(row.get('Consol Days Full')),
        'range_pct'       : safe_round(row.get('Consol Range'),    1),
        'range_high'      : safe_round(row.get('Consol High'),     2),
        'range_low'       : safe_round(row.get('Consol Low'),      2),
        'pct_to_breakout' : safe_round(row.get('Pct to Breakout'), 1),
        'volume'          : safe_round(row.get('Breakout Vol'),    1),
    }

    card = {
        'symbol'      : symbol,
        'sector'      : str(row.get('Sector', '')),
        'price'       : safe_round(row.get('Current Price'), 2),
        'tier'        : str(row.get('Tier', '')),
        'best_setup'  : str(row.get('Best Setup', '')),
        'tech_score'  : safe_int(row.get('Tech Score')),
        'fund_score'  : safe_round(row.get('Final Score'), 1),
        'sector_score': safe_round(row.get('Sector Score'), 1),
        'hist_score'  : safe_round(row.get('Historical Score'), 1),
        'peer_score'  : safe_round(row.get('Peer Score'), 1),
        'quality_score': safe_round(row.get('Quality Score'), 1),
        'entry'       : safe_round(row.get('Entry'), 2),
        'stop_loss'   : safe_round(row.get('Stop Loss'), 2),
        'target'      : safe_round(row.get('Target'), 2),
        'fundamentals': fund_details,
        'technicals'  : tech_details,
        'consolidation': consol_details,
    }

    stock_cards.append(card)

with open('../data/stock_cards.json', 'w') as f:
    json.dump(stock_cards, f, indent=2)

print(f"Stock cards created: {len(stock_cards)} ✅")
print(f"Saved to data/stock_cards.json ✅")

# Show sector scores
print(f"\n=== SECTOR RELATIVE SCORES ===")
for sector in sorted(fund_df['Sector'].unique()):
    sector_stocks = fund_df[fund_df['Sector'] == sector][
        ['Symbol', 'Final Score', 'Sector Score']
    ].sort_values('Sector Score', ascending=False)
    print(f"\n{sector}:")
    for _, r in sector_stocks.iterrows():
        bar = '█' * int(r['Sector Score'])
        print(f"  {r['Symbol']:15} "
              f"Fund:{r['Final Score']:5.1f}  "
              f"Sector:{r['Sector Score']:4.1f}/10  "
              f"{bar}")
```

```python
# Regenerate with updated tier logic
print("Regenerating with updated tier classification...")

tech_reports = []

for _, fund_row in fundamental_df.iterrows():
    symbol     = fund_row['Symbol']
    fund_score = fund_row['Final Score']
    sector     = fund_row['Sector']

    if symbol not in indicator_data:
        continue

    df       = indicator_data[symbol]
    latest   = df.iloc[-1]

    close    = latest['Close']
    rsi      = round(latest['RSI'],       2)
    adx      = round(latest['ADX'],       2)
    hist     = round(latest['MACD_Hist'], 2)
    vol_r    = round(latest['Vol_Ratio'], 2)
    ema20    = round(latest['EMA20'],     2)
    ema50    = round(latest['EMA50'],     2)
    ema200   = round(latest['EMA200'],    2)
    di_plus  = round(latest['DI_Plus'],   2)
    di_minus = round(latest['DI_Minus'],  2)

    vp           = calculate_volume_profile(df)
    mom_score, _ = score_momentum(df, vp)
    rev_score, _ = score_reversal(df, vp)
    consol_info  = get_consolidation_info(df)

    if mom_score >= rev_score and mom_score >= 50:
        best_setup = 'Momentum'
        tech_score = mom_score
    elif rev_score >= mom_score and rev_score >= 50:
        best_setup = 'Reversal'
        tech_score = rev_score
    else:
        best_setup = 'Watching'
        tech_score = max(mom_score, rev_score)

    in_consol       = consol_info['consolidating']
    consol_days     = consol_info['consol_days']
    consol_label    = consol_info['duration_label']
    pct_to_breakout = consol_info['pct_to_breakout'] or 0
    consol_volume   = consol_info['breakout_volume']  or 0

    # ── UPDATED TIER LOGIC ──
    if fund_score >= 60 and tech_score >= 65 and best_setup != 'Watching':
        if best_setup == 'Momentum':
            tier = 'TIER 1 — BUY NOW (Momentum)'
        else:
            tier = 'TIER 1 — BUY NOW (Reversal)'
    elif fund_score >= 60 and in_consol and pct_to_breakout > -5:
        tier = 'TIER 1 — BREAKOUT IMMINENT'
    elif fund_score >= 60 and tech_score >= 40 and best_setup != 'Watching':
        tier = 'TIER 2 — WATCHLIST'
    elif fund_score >= 60 and in_consol:
        tier = 'TIER 2 — BASE BUILDING'
    else:
        tier = 'TIER 3 — WAITING'

    if vp:
        poc = vp['poc_price']
        vah = vp['vah']
        val = vp['val']
    else:
        poc = vah = val = None

    if best_setup == 'Momentum':
        entry  = close
        stop   = round(ema50  * 0.97, 2)
        target = round(close  * 1.20, 2)
    elif best_setup == 'Reversal':
        entry  = close
        stop   = round(df['Low'].tail(20).min() * 0.97, 2)
        target = round(ema200 * 1.05, 2)
    elif in_consol and consol_info['range_high']:
        entry  = round(consol_info['range_high'] * 1.02, 2)
        stop   = round(consol_info['range_low']  * 0.97, 2)
        target = round(consol_info['range_high'] +
                  (consol_info['range_high'] -
                   consol_info['range_low']) * 0.75, 2)
    else:
        entry = stop = target = None

    tech_reports.append({
        'Symbol'          : symbol,
        'Sector'          : sector,
        'Fund Score'      : fund_score,
        'Current Price'   : round(close,   2),
        'RSI'             : rsi,
        'ADX'             : adx,
        'MACD Hist'       : hist,
        'Vol Ratio'       : vol_r,
        'EMA20'           : ema20,
        'EMA50'           : ema50,
        'EMA200'          : ema200,
        'DI+'             : di_plus,
        'DI-'             : di_minus,
        'Momentum Score'  : mom_score,
        'Reversal Score'  : rev_score,
        'Best Setup'      : best_setup,
        'Tech Score'      : tech_score,
        'Tier'            : tier,
        'In Consolidation': in_consol,
        'Consol Days'     : consol_days,
        'Consol Label'    : consol_label,
        'Pct to Breakout' : round(pct_to_breakout, 2),
        'Consol Volume'   : consol_volume,
        'Entry'           : entry,
        'Stop Loss'       : stop,
        'Target'          : target,
        'POC'             : poc,
        'VAH'             : vah,
        'VAL'             : val,
    })

tech_report_df = pd.DataFrame(tech_reports)

# Add sector scores
fund_df['Sector Score'] = 0.0
for sector in fund_df['Sector'].unique():
    mask      = fund_df['Sector'] == sector
    max_score = fund_df[mask]['Final Score'].max()
    fund_df.loc[mask, 'Sector Score'] = (
        fund_df[mask]['Final Score'] / max_score * 10
    ).round(1)

tech_report_df = tech_report_df.merge(
    fund_df[['Symbol', 'Sector Score']],
    on='Symbol', how='left'
)

tech_report_df.to_csv('../data/technical_report.csv', index=False)
print(f"Report regenerated: {len(tech_report_df)} stocks")
print(f"\nTier distribution:")
print(tech_report_df['Tier'].value_counts().to_string())
```
