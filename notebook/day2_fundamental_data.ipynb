```python
def get_fundamental_data(symbol):
    """
    Fetch complete fundamental data for any NSE stock
    symbol: NSE symbol like 'RELIANCE', 'TCS', 'INFY'
    """
    try:
        # Add .NS for NSE
        ticker = yf.Ticker(f"{symbol}.NS")
        info   = ticker.info
        
        # Check if valid stock
        if 'longName' not in info:
            return None
        
        # Financial statements
        financials    = ticker.financials
        balance_sheet = ticker.balance_sheet
        cashflow      = ticker.cashflow
        
        # Skip if insufficient data
        if financials.empty or balance_sheet.empty or cashflow.empty:
            return None
        
        # --- CALCULATE METRICS ---
        
        # ROE
        try:
            net_income          = financials.loc['Net Income'].iloc[0]
            stockholders_equity = balance_sheet.loc['Stockholders Equity'].iloc[0]
            roe                 = net_income / stockholders_equity
        except:
            roe = None
            
        # Free Cash Flow
        try:
            free_cashflow = cashflow.loc['Free Cash Flow'].iloc[0]
        except:
            free_cashflow = None
            
        # Operating Cash Flow
        try:
            operating_cf = cashflow.loc['Operating Cash Flow'].iloc[0]
        except:
            operating_cf = None
        
        # Revenue Growth (YoY)
        try:
            revenues       = financials.loc['Total Revenue']
            revenue_growth = (revenues.iloc[0] - revenues.iloc[1]) / revenues.iloc[1]
        except:
            revenue_growth = None
            
        # Profit Growth (YoY)
        try:
            profits       = financials.loc['Net Income']
            profit_growth = (profits.iloc[0] - profits.iloc[1]) / profits.iloc[1]
        except:
            profit_growth = None
        
        # --- BUILD OUTPUT ---
        data = {
            'Symbol'              : symbol,
            'Company'             : info.get('longName'),
            'Sector'              : info.get('sector'),
            'Industry'            : info.get('industry'),
            'Market Cap'          : info.get('marketCap'),
            'Current Price'       : info.get('currentPrice'),
            
            # Valuation
            'PE Ratio'            : info.get('trailingPE'),
            'Price to Book'       : info.get('priceToBook'),
            'Price to Sales'      : info.get('priceToSalesTrailing12Months'),
            
            # Profitability
            'ROE'                 : roe,
            'Profit Margins'      : info.get('profitMargins'),
            'Operating Margins'   : info.get('operatingMargins'),
            'Gross Margins'       : info.get('grossMargins'),
            
            # Growth
            'Revenue Growth'      : revenue_growth,
            'Profit Growth'       : profit_growth,
            'Earnings Growth'     : info.get('earningsGrowth'),
            
            # Financial Health
            'Debt to Equity'      : info.get('debtToEquity'),
            'Free Cash Flow'      : free_cashflow,
            'Operating CF'        : operating_cf,
            
            # Ownership
            'Promoter Holding'    : info.get('heldPercentInsiders'),
            'Institution Holding' : info.get('heldPercentInstitutions'),
            
            # Analyst
            'Recommendation'      : info.get('recommendationKey'),
            'Current Price'       : info.get('currentPrice'),
            'Target Price'        : info.get('targetMedianPrice'),
        }
        
        return data
    
    except Exception as e:
        print(f"Error fetching {symbol}: {e}")
        return None

print("Function defined successfully!")
```

```python
# Test on 5 stocks across different market caps
test_stocks = [
    'TCS',        # Large cap - IT
    'PERSISTENT', # Mid cap - IT
    'HAPPSTMNDS', # Small cap - IT
    'IRFC',       # Mid cap - Finance
    'BBOX',       # Small cap
]

results = []

for symbol in test_stocks:
    print(f"Fetching {symbol}...")
    data = get_fundamental_data(symbol)
    if data:
        results.append(data)
        print(f"✓ {symbol} done")
    else:
        print(f"✗ {symbol} failed")

print(f"\nSuccessfully fetched: {len(results)}/{len(test_stocks)} stocks")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

def scrape_screener(symbol):
    """
    Scrape quarterly financial data from Screener.in
    symbol: NSE symbol like 'RELIANCE', 'TCS'
    """
    try:
        url = f"https://www.screener.in/company/{symbol}/consolidated/"
        
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }
        
        response = requests.get(url, headers=headers, timeout=10)
        print(f"Status code: {response.status_code}")
        
        if response.status_code != 200:
            print(f"Failed to fetch {symbol}")
            return None
            
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find quarterly results section
        sections = soup.find_all('section')
        print(f"Total sections found: {len(sections)}")
        
        for section in sections:
            heading = section.find('h2')
            if heading:
                print(f"Section: {heading.text.strip()}")
        
        return soup
    
    except Exception as e:
        print(f"Error: {e}")
        return None

# Test with Reliance
soup = scrape_screener('RELIANCE')
```

```python
def get_quarterly_data(symbol):
    try:
        url = f"https://www.screener.in/company/{symbol}/consolidated/"
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }
        
        response = requests.get(url, headers=headers, timeout=10)
        soup     = BeautifulSoup(response.text, 'html.parser')
        
        # Find quarterly results section
        quarterly_section = None
        for section in soup.find_all('section'):
            heading = section.find('h2')
            if heading and 'Quarterly Results' in heading.text:
                quarterly_section = section
                break
        
        if not quarterly_section:
            return None
        
        table = quarterly_section.find('table')
        
        # Extract column headers (quarter names)
        headers_row = table.find('thead').find_all('th')
        columns     = []
        for th in headers_row:
            text = th.get_text(strip=True)
            columns.append(text if text else 'Metric')
        
        # Extract data rows
        rows     = table.find('tbody').find_all('tr')
        data     = {}
        
        for row in rows:
            cells      = row.find_all('td')
            if not cells:
                continue
            
            # First cell is metric name
            metric_name = cells[0].get_text(strip=True)
            
            # Skip Raw PDF row
            if 'Raw PDF' in metric_name or 'PDF' in metric_name:
                continue
            
            # Clean metric name (remove + signs)
            metric_name = metric_name.replace('+', '').strip()
            
            # Remaining cells are values
            values = []
            for cell in cells[1:]:
                val = cell.get_text(strip=True)
                val = val.replace(',', '').replace('%', '').strip()
                try:
                    values.append(float(val))
                except:
                    values.append(None)
            
            data[metric_name] = values
        
        # Create dataframe
        quarter_cols = columns[1:]  # Remove 'Metric' column
        df           = pd.DataFrame(data, index=quarter_cols).T
        
        print(f"✓ {symbol} — {df.shape[1]} quarters, {df.shape[0]} metrics")
        print(df.index.tolist())
        return df
    
    except Exception as e:
        print(f"Error: {e}")
        return None

# Test
quarterly_df = get_quarterly_data('RELIANCE')
quarterly_df
```

```python
def scrape_screener_complete(symbol):
    """
    Extract all sections from Screener.in for a stock
    Returns dictionary with all sections as dataframes
    """
    try:
        url      = f"https://www.screener.in/company/{symbol}/consolidated/"
        headers  = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }
        
        response = requests.get(url, headers=headers, timeout=10)
        
        if response.status_code != 200:
            print(f"Failed: {symbol} — status {response.status_code}")
            return None
            
        soup     = BeautifulSoup(response.text, 'html.parser')
        sections = soup.find_all('section')
        
        # Sections we want to extract
        target_sections = [
            'Quarterly Results',
            'Profit & Loss',
            'Balance Sheet',
            'Cash Flows',
            'Ratios',
            'Shareholding Pattern',
            'Peer Comparison',
        ]
        
        result = {}
        
        for section in sections:
            heading = section.find('h2')
            if not heading:
                continue
                
            section_name = heading.text.strip()
            
            if section_name not in target_sections:
                continue
            
            # Extract table from section
            table = section.find('table')
            if not table:
                continue
            
            df = extract_table(table, section_name)
            
            if df is not None:
                result[section_name] = df
                print(f"  ✓ {section_name}: {df.shape[0]} rows × {df.shape[1]} cols")
        
        return result
    
    except Exception as e:
        print(f"Error scraping {symbol}: {e}")
        return None


def extract_table(table, section_name):
    """
    Extract any table from Screener.in cleanly
    """
    try:
        # Get column headers
        thead    = table.find('thead')
        columns  = []
        
        if thead:
            for th in thead.find_all('th'):
                text = th.get_text(strip=True)
                columns.append(text if text else 'Metric')
        
        # Get data rows
        tbody = table.find('tbody')
        if not tbody:
            return None
            
        rows = tbody.find_all('tr')
        data = {}
        
        for row in rows:
            cells = row.find_all('td')
            if not cells:
                continue
            
            # First cell = metric name
            metric_name = cells[0].get_text(strip=True)
            metric_name = metric_name.replace('+', '').strip()
            
            # Skip unwanted rows
            skip_keywords = ['Raw PDF', 'PDF', 'Source']
            if any(kw in metric_name for kw in skip_keywords):
                continue
            
            # Extract values
            values = []
            for cell in cells[1:]:
                val = cell.get_text(strip=True)
                val = val.replace(',', '').replace('%', '').strip()
                try:
                    values.append(float(val))
                except:
                    # Keep as string if not numeric
                    values.append(val if val else None)
            
            data[metric_name] = values
        
        if not data:
            return None
        
        # Create dataframe
        col_names = columns[1:] if len(columns) > 1 else list(range(len(values)))
        df        = pd.DataFrame(data, index=col_names).T
        
        return df
    
    except Exception as e:
        print(f"Table extraction error: {e}")
        return None


# Test with Reliance
print("Fetching complete data for RELIANCE...")
reliance_data = scrape_screener_complete('RELIANCE')

print(f"\nSections successfully extracted: {list(reliance_data.keys())}")
```
